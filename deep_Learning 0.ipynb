##Introduction




ก่อนหน้านี้ โปรแกรมคอมพิวเตอร์เกือบทุกโปรแกรมที่คุณอาจเคยโต้ตอบด้วยในระหว่างวันปกติล้วนได้รับการเข้ารหัสเป็นชุดกฎที่เข้มงวดที่ระบุอย่างแม่นยำว่าควรทำงานอย่างไร ตัวอย่างเช่น สมมติว่าเราต้องการเขียนแอปพลิเคชันเพื่อจัดการแพลตฟอร์มอีคอมเมิร์ซ หลังจากประชุมรอบกระดานไวท์บอร์ดเพื่อพิจารณาปัญหาเป็นเวลาหลายชั่วโมง เราอาจตัดสินใจเกี่ยวกับวิธีแก้ปัญหาการทำงานในภาพกว้าง เช่น: (i) ผู้ใช้โต้ตอบกับแอปพลิเคชันผ่านอินเทอร์เฟซที่ทำงานในเว็บเบราว์เซอร์หรือแอปพลิเคชันมือถือ (ii) แอปพลิเคชันของเราโต้ตอบกับเครื่องมือจัดการฐานข้อมูลเชิงพาณิชย์เพื่อติดตามสถานะของผู้ใช้แต่ละคนและเก็บรักษาบันทึกธุรกรรมในอดีต และ (iii) ที่ใจกลางแอปพลิเคชันของเรา ลอจิกทางธุรกิจ (คุณอาจพูดว่า สมอง) ของแอปพลิเคชันของเราจะกำหนดชุดกฎที่จับคู่สถานการณ์ที่อาจเกิดขึ้นได้ทั้งหมดกับการกระทำที่สอดคล้องกันที่โปรแกรมของเราควรดำเนินการ

เพื่อสร้างสมองของแอปพลิเคชันของเรา เราอาจระบุเหตุการณ์ทั่วไปทั้งหมดที่โปรแกรมของเราควรจัดการ ตัวอย่างเช่น เมื่อใดก็ตามที่ลูกค้าคลิกเพื่อเพิ่มสินค้าลงในรถเข็น โปรแกรมของเราควรเพิ่มรายการลงในตารางฐานข้อมูลรถเข็น ซึ่งเชื่อมโยง ID ของผู้ใช้กับ ID ของผลิตภัณฑ์ที่ร้องขอ เราอาจพยายามก้าวผ่านกรณีที่เป็นไปได้ทั้งหมด ทดสอบความเหมาะสมของกฎของเรา และทำการแก้ไขที่จำเป็น สิ่งที่จะเกิดขึ้นหากผู้ใช้เริ่มต้นการซื้อโดยไม่มีสินค้าในรถเข็น แม้ว่านักพัฒนาไม่กี่คนจะทำได้ถูกต้องตั้งแต่ครั้งแรก (อาจต้องใช้การทดสอบบางครั้งเพื่อให้ได้ผลลัพธ์ที่ดีที่สุด) แต่ส่วนใหญ่แล้ว เราสามารถเขียนโปรแกรมดังกล่าวและเปิดตัวได้อย่างมั่นใจโดยไม่เคยเห็นลูกค้าจริง ๆ ความสามารถของเราในการออกแบบระบบอัตโนมัติด้วยตนเองที่ขับเคลื่อนผลิตภัณฑ์และระบบที่ใช้งานได้จริง ในสถานการณ์ใหม่ ๆ มักเป็นความสามารถทางปัญญาที่น่าทึ่ง และเมื่อคุณสามารถออกแบบโซลูชันที่ทำงานได้ 100% ของเวลา คุณไม่ควรต้องกังวลเกี่ยวกับการเรียนรู้ของเครื่อง

โชคดีสำหรับชุมชนของนักวิทยาศาสตร์การเรียนรู้ของเครื่องที่กำลังเติบโตขึ้น หลายงานที่เราต้องการทำให้อัตโนมัติไม่สามารถทำได้ง่ายดายเพียงนี้ด้วยความคิดสร้างสรรค์ของมนุษย์ ลองนึกภาพว่าการประชุมรอบกระดานไวท์บอร์ดกับจิตใจที่ชาญฉลาดที่สุดที่คุณรู้จัก แต่ในครั้งนี้คุณกำลังแก้ไขปัญหาต่อไปนี้:

เขียนโปรแกรมที่คาดการณ์สภาพอากาศในวันพรุ่งนี้โดยพิจารณาจากข้อมูลทางภูมิศาสตร์ ภาพถ่ายจากดาวเทียม และข้อมูลสภาพอากาศในอดีต
เขียนโปรแกรมที่รับคำถามข้อเท็จจริงที่แสดงเป็นข้อความแบบฟรีฟอร์มและตอบให้ถูกต้อง
เขียนโปรแกรมที่ระบุบุคคลทุกคนที่ปรากฏในรูปภาพและวาดเส้นรอบ ๆ แต่ละคน
ปัญหาเหล่านี้มีความท้าทายเพราะมีกฎเกณฑ์ที่ชัดเจนเพียงเล็กน้อยหรือไม่มีเลยที่ระบุว่าโปรแกรมควรทำงานอย่างไรในสถานการณ์ที่เป็นไปได้ทั้งหมด ด้วยเหตุนี้เราจึงหันมาใช้การเรียนรู้ของเครื่องเพื่อช่วยเราในการสร้างโปรแกรมที่สามารถแก้ปัญหาเหล่านี้ได้

การเรียนรู้ของเครื่องเป็นสาขาหนึ่งของวิทยาการคอมพิวเตอร์ที่เกี่ยวข้องกับการพัฒนาอัลกอริทึมที่สามารถเรียนรู้จากข้อมูลและปรับปรุงประสิทธิภาพของงานที่กำหนดไว้โดยไม่ได้รับการเขียนโปรแกรมอย่างชัดเจน อัลกอริทึมการเรียนรู้ของเครื่อง



* เขียนโปรแกรมที่แสดงสินค้าที่ผู้ใช้มีแนวโน้มที่จะชอบแต่ไม่น่าจะพบเจอในขณะเรียกดูตามปกติ

สำหรับปัญหาเหล่านี้ แม้แต่โปรแกรมเมอร์ชั้นยอดก็ยังยากที่จะเขียนโปรแกรมแก้ปัญหาจากศูนย์ เหตุผลอาจแตกต่างกันไป บางครั้งโปรแกรมที่เรากำลังมองหาอาจเปลี่ยนไปตามเวลา ดังนั้นจึงไม่มีคำตอบที่ถูกต้องตายตัว ในกรณีเช่นนี้ โซลูชันใดๆ ที่ประสบความสำเร็จจะต้องปรับตัวให้เข้ากับโลกที่เปลี่ยนแปลงอย่างราบรื่น บางครั้งความสัมพันธ์ (เช่น ระหว่างพิกเซลกับหมวดหมู่แบบนามธรรม) อาจซับซ้อนเกินไป ต้องใช้การคำนวณหลายพันหรือหลายล้านครั้งและปฏิบัติตามหลักการที่ไม่รู้จัก ในกรณีของการจดจำภาพ ขั้นตอนที่แม่นยำที่จำเป็นในการปฏิบัติงานนั้นอยู่นอกเหนือความเข้าใจอย่างมีสติของเรา แม้ว่ากระบวนการทางปัญญาแบบไม่รู้ตัวของเราจะดำเนินการงานนั้นได้อย่างง่ายดาย

การเรียนรู้ของเครื่องคือการศึกษาอัลกอริธึมที่สามารถเรียนรู้จากประสบการณ์ เมื่ออัลกอริธึมการเรียนรู้ของเครื่องได้รับประสบการณ์มากขึ้น โดยทั่วไปในรูปแบบของข้อมูลการสังเกตหรือการโต้ตอบกับสภาพแวดล้อม ประสิทธิภาพของมันจะดีขึ้น เปรียบเทียบสิ่งนี้กับแพลตฟอร์มอีคอมเมิร์ซแบบกำหนดของเรา ซึ่งปฏิบัติตามลอจิกทางธุรกิจเดียวกัน ไม่ว่าจะได้รับประสบการณ์มากแค่ไหน จนกว่านักพัฒนาเองจะเรียนรู้และตัดสินใจว่าถึงเวลาอัปเดตซอฟต์แวร์แล้ว ในหนังสือเล่มนี้ เราจะสอนคุณเกี่ยวกับพื้นฐานของการเรียนรู้ของเครื่อง โดยเฉพาะอย่างยิ่งเกี่ยวกับการเรียนรู้เชิงลึก ซึ่งเป็นชุดเทคนิคที่มีประสิทธิภาพที่ขับเคลื่อนการริเริ่มในสาขาต่างๆ เช่น คอมพิวเตอร์วิชัน การประมวลผลภาษาธรรมชาติ การดูแลสุขภาพ และจีโนมิกส์
