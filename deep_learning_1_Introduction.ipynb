## 1.1 A Motivating Example

# 1.2 Key Components




ในตัวอย่างคำสั่งปลุกของเรา เราได้อธิบายถึงชุดข้อมูลที่ประกอบด้วยชิ้นส่วนเสียงและป้ายกำกับแบบไบนารี และเราได้ให้ความรู้สึกคร่าวๆ ว่าเราจะฝึกโมเดลเพื่อประมาณการแมปจากชิ้นส่วนไปยังการจัดประเภทได้อย่างไร ปัญหาประเภทนี้ ซึ่งเราพยายามคาดการณ์ป้ายกำกับที่ไม่รู้จักตามกำหนดไว้ โดยอิงจากข้อมูลเข้าที่ทราบแล้วจากชุดข้อมูลที่มีตัวอย่างที่ทราบป้ายกำกับ เรียกว่าการเรียนรู้แบบมีผู้สอน (supervised learning) นี่เป็นเพียงหนึ่งในหลายประเภทของปัญหาการเรียนรู้ของเครื่อง ก่อนที่เราจะสำรวจประเภทอื่นๆ เราต้องการอธิบายเพิ่มเติมเกี่ยวกับส่วนประกอบหลักบางอย่างที่เราจะใช้ ไม่ว่าเราจะแก้ปัญหาการเรียนรู้ของเครื่องประเภทใดก็ตาม

1. ข้อมูลที่เราสามารถเรียนรู้ได้
2. โมเดลของวิธีแปลงข้อมูล
3. ฟังก์ชันวัตถุประสงค์ที่วัดผลว่าโมเดลทำได้ดี (หรือแย่) เพียงใด
4. อัลกอริทึมสำหรับปรับพารามิเตอร์ของโมเดลเพื่อเพิ่มประสิทธิภาพฟังก์ชันวัตถุประสงค์

**คำอธิบายเพิ่มเติม**

* ข้อมูลที่เราสามารถเรียนรู้ได้ เป็นสิ่งที่จำเป็นสำหรับการฝึกโมเดลการเรียนรู้ของเครื่อง ข้อมูลสามารถมีรูปแบบต่างๆ เช่น ข้อความ อิมเมจ เสียง หรือข้อมูลเซ็นเซอร์
* โมเดลของวิธีแปลงข้อมูล เป็นวิธีที่เราใช้ในการแปลงข้อมูลดิบเป็นรูปแบบที่คอมพิวเตอร์สามารถเข้าใจได้ โมเดลสามารถเป็นอะไรก็ได้ เช่น ตารางการตัดสินใจ ฟังก์ชันคณิตศาสตร์ หรือเครือข่ายประสาทเทียม
* ฟังก์ชันวัตถุประสงค์ เป็นวิธีที่เราใช้ในการวัดประสิทธิภาพของโมเดล ฟังก์ชันวัตถุประสงค์จะคำนวณค่าความผิดพลาดของโมเดล ซึ่งเป็นตัวบ่งชี้ว่าโมเดลคาดการณ์ได้ถูกต้องเพียงใด
* อัลกอริทึมสำหรับปรับพารามิเตอร์ของโมเดลเพื่อเพิ่มประสิทธิภาพฟังก์ชันวัตถุประสงค์ เป็นวิธีที่เราใช้ในการปรับปรุงโมเดล อัลกอริทึมจะทำงานโดยปรับพารามิเตอร์ของโมเดลจนกว่าฟังก์ชันวัตถุประสงค์จะลดลง ซึ่งหมายความว่าโมเดลมีความแม่นยำมากขึ้น

**ตัวอย่าง**

ให้พิจารณาปัญหาการจำแนกประเภทภาพ โดยที่เราต้องการฝึกโมเดลเพื่อแยกแยะระหว่างภาพแมวและสุนัข ในกรณีนี้ ข้อมูลที่เราสามารถเรียนรู้ได้ จะเป็นชุดข้อมูลของภาพแมวและสุนัขที่มีป้ายกำกับ โมเดลของวิธีแปลงข้อมูล อาจเป็นเครือข่ายประสาทเทียมที่ได้รับการฝึกฝนเพื่อดึงดูดคุณลักษณะที่สำคัญจากภาพ ฟังก์ชันวัตถุประสงค์ อาจเป็นการสูญเสียฟังก์ชันที่คำนวณความแตกต่างระหว่างการคาดการณ์ของโมเดลและป้ายกำกับที่แท้จริงของข้อมูล และอัลกอริทึมสำหรับปรับพารามิเตอร์ของโมเดลเพื่อเพิ่มประสิทธิภาพฟังก์ชันวัตถุประสงค์ อาจเป็นอัลกอริทึมการแพร่กระจายย้อนกลับ (backpropagation)

ส่วนประกอบหลักทั้งสี่นี้มีความสำคัญต่อการเรียนรู้ของเครื่องทั้งหมด ไม่ว่าเราจะแก้ปัญหาประเภทใดก็ตาม โดยการเข้าใจส่วนประกอบเหล่านี้ เราจะสามารถพัฒนาโมเดลการเรียนรู้ของเครื่องที่มีประสิทธิภาพมากขึ้น

 1.2.2 Models



การเรียนรู้ของเครื่องส่วนใหญ่เกี่ยวข้องกับการแปลงข้อมูลในบางแง่มุม เราอาจต้องการสร้างระบบที่รับรูปภาพและคาดการณ์ว่าเป็นรูปภาพที่ยิ้มหรือไม่ หรือเราอาจต้องการรับชุดค่าอ่านจากเซ็นเซอร์และคาดการณ์ว่าค่าอ่านเหล่านั้นเป็นปกติหรือผิดปกติ โดยคำว่า โมเดล (model) เราหมายถึงกลไกการคำนวณสำหรับการรับข้อมูลประเภทหนึ่งและส่งผลลัพธ์เป็นการคาดการณ์ที่มีประเภทอาจแตกต่างออกไป โดยเฉพาะอย่างยิ่ง เราสนใจในโมเดลสถิติที่สามารถประมาณการจากข้อมูลได้ แม้ว่าโมเดลที่เรียบง่ายจะสามารถแก้ไขปัญหาที่เรียบง่ายได้อย่างสมบูรณ์แบบ แต่ปัญหาที่เราจะมุ่งเน้นในหนังสือเล่มนี้จะทดสอบขีดจำกัดของวิธีการแบบดั้งเดิม การเรียนรู้เชิงลึกแตกต่างจากวิธีการแบบดั้งเดิมโดยหลักสำคัญที่ชุดของโมเดลที่มีประสิทธิภาพที่มันมุ่งเน้นไปที่ โมเดลเหล่านี้ประกอบด้วยการแปลงข้อมูลหลายครั้งซ้อนกัน ซึ่งเชื่อมโยงกันจากบนลงล่าง ดังนั้นจึงใช้ชื่อว่าการเรียนรู้เชิงลึก ในระหว่างทางไปสู่การอภิปรายโมเดลเชิงลึก เราจะอภิปรายวิธีการแบบดั้งเดิมบางอย่างเช่นกัน

**คำอธิบายเพิ่มเติม**

* การแปลงข้อมูล เป็นกระบวนการเปลี่ยนข้อมูลจากรูปแบบหนึ่งไปเป็นอีกรูปแบบหนึ่ง ในการเรียนรู้ของเครื่อง เราอาจต้องการแปลงข้อมูลเพื่อให้คอมพิวเตอร์เข้าใจได้ง่ายขึ้น หรือเพื่อให้ได้ข้อมูลที่มีประโยชน์มากขึ้น
* โมเดลสถิติ เป็นโมเดลที่ใช้สถิติเพื่ออธิบายความสัมพันธ์ระหว่างข้อมูลที่แตกต่างกัน โมเดลสถิติสามารถใช้เพื่อคาดการณ์เหตุการณ์ในอนาคตหรือเพื่อทำความเข้าใจเกี่ยวกับข้อมูลที่มีอยู่
* การเรียนรู้เชิงลึก เป็นประเภทหนึ่งของการเรียนรู้ของเครื่องที่ใช้เครือข่ายประสาทเทียม เครือข่ายประสาทเทียมเป็นโมเดลคณิตศาสตร์ที่ประกอบด้วยชั้นของหน่วยประมวลผลที่เชื่อมต่อกัน ซึ่งสามารถเรียนรู้จากข้อมูลและทำการคาดการณ์
* การแปลงข้อมูลหลายครั้งซ้อนกัน เป็นเทคนิคที่ใช้ในเครือข่ายประสาทเทียม โดยแต่ละชั้นของเครือข่ายจะแปลงข้อมูลจากชั้นก่อนหน้าและส่งต่อไปยังชั้นถัดไป
* วิธีการแบบดั้งเดิม เป็นวิธีการเรียนรู้ของเครื่องที่ไม่ใช่การเรียนรู้เชิงลึก เช่น เครื่องตัดสินใจต้นไม้ (decision tree) และการสนับสนุนเวกเตอร์แมชชีน (support vector machine)

**ตัวอย่าง**

ให้พิจารณาปัญหาการจำแนกประเภทภาพอีกครั้ง ในกรณีนี้ เราอาจใช้เครือข่ายประสาทเทียมเป็นโมเดล เครือข่ายประสาทเทียมจะแปลงรูปภาพเป็นชุดของคุณลักษณะที่สำคัญก่อนที่จะทำการคาดการณ์ว่ารูปภาพนั้นเป็นภาพแมวหรือภาพสุนัข การแปลงข้อมูลหลายครั้งซ้อนกันจะช่วยให้เครือข่ายประสาทเทียมเรียนรู้คุณลักษณะที่ซับซ้อนมากขึ้นได้

การเรียนรู้เชิงลึกเป็นวิธีการเรียนรู้ของเครื่องที่ทรงพลังมาก แต่ก็มีความซับซ้อนสูงเช่นกัน ในหนังสือเล่มนี้ เราจะอธิบายพื้นฐานของการเรียนรู้เชิงลึกและวิธีใช้เครือข่ายประสาทเทียมเพื่อแก้ปัญหาการเรียนรู้ของเครื่องประเภทต่างๆ

1.2.3 Objective Functions

**การแปลและเรียบเรียงคำเหมือนตำรา**

ก่อนหน้านี้ เราได้แนะนำการเรียนรู้ของเครื่องว่าเป็นการเรียนรู้จากประสบการณ์ การเรียนรู้ในที่นี้ หมายถึงการปรับปรุงประสิทธิภาพในการทำงานบางอย่างเมื่อเวลาผ่านไป แต่ใครจะเป็นผู้ตัดสินว่าอะไรคือการปรับปรุง? คุณอาจคิดว่าเราสามารถเสนอที่จะอัปเดตโมเดลของเราได้ และบางคนอาจไม่เห็นด้วยว่าข้อเสนอของเราเป็นการปรับปรุงหรือไม่

เพื่อพัฒนาระบบการเรียนรู้ของเครื่องอย่างเป็นทางการ เราจำเป็นต้องมีการวัดอย่างเป็นทางการว่าโมเดลของเราดี (หรือแย่) เพียงใด ในการเรียนรู้ของเครื่อง และการหาค่าที่ดีที่สุดโดยทั่วไป เราเรียกสิ่งเหล่านี้ว่า ฟังก์ชันวัตถุประสงค์ (objective function) ตามธรรมเนียม เราจะกำหนดฟังก์ชันวัตถุประสงค์ให้ต่ำเท่าไหร่ยิ่งดี นี่เป็นเพียงแค่ธรรมเนียมเท่านั้น คุณสามารถใช้ฟังก์ชันใดก็ได้ที่ยิ่งสูงยิ่งดี และเปลี่ยนเป็นฟังก์ชันใหม่ที่มีคุณภาพเหมือนกันแต่ต่ำเท่าไหร่ยิ่งดีได้โดยการพลิกเครื่องหมาย เนื่องจากเราเลือกให้ต่ำคือดีกว่า ฟังก์ชันเหล่านี้บางครั้งจึงเรียกว่า ฟังก์ชันการสูญเสีย (loss function)

เมื่อพยายามทำนายค่าตัวเลข ฟังก์ชันการสูญเสียที่พบบ่อยที่สุดคือข้อผิดพลาดแบบกำลังสอง (squared error) ซึ่งคือกำลังสองของความแตกต่างระหว่างการคาดการณ์และเป้าหมายจริง สำหรับการจำแนกประเภท วัตถุประสงค์ที่พบบ่อยที่สุดคือการลดอัตราข้อผิดพลาด (error rate) ซึ่งคือสัดส่วนของตัวอย่างที่การคาดการณ์ของเราไม่ตรงกับเป้าหมายจริง วัตถุประสงค์บางอย่าง (เช่น ข้อผิดพลาดแบบกำลังสอง) ง่ายต่อการหาค่าที่ดีที่สุด ในขณะที่บางอย่าง (เช่น อัตราข้อผิดพลาด) ยากต่อการหาค่าที่ดีที่สุดโดยตรง เนื่องจากไม่สามารถหาอนุพันธ์ได้หรือมีความซับซ้อนอื่นๆ ในกรณีเหล่านี้ เป็นเรื่องปกติที่จะใช้การหาค่าที่ดีที่สุดของวัตถุประสงค์ตัวแทน (surrogate objective) แทน

ระหว่างการหาค่าที่ดีที่สุด เราคิดว่าการสูญเสียเป็นฟังก์ชันของพารามิเตอร์ของโมเดล และปฏิบัติต่อชุดข้อมูลฝึกอบรมเป็นค่าคงที่ เราเรียนรู้ค่าที่ดีที่สุดของพารามิเตอร์โมเดลของเราโดยการลดการสูญเสียที่เกิดขึ้นบนชุดที่มีตัวอย่างจำนวนหนึ่งที่รวบรวมมาเพื่อการฝึกอบรม อย่างไรก็ตาม การทำได้ดีกับข้อมูลฝึกอบรมไม่ได้รับประกันว่าเราจะทำได้ดีกับข้อมูลที่ไม่เคยเห็นมาก่อน ดังนั้น เราจึงมักต้องการแบ่งข้อมูลที่มีอยู่เป็นสองส่วน: ชุดข้อมูลฝึกอบรม (หรือชุดฝึกอบรม) สำหรับการเรียนรู้พารามิเตอร์โมเดล และชุดข้อมูลทดสอบ (หรือชุดทดสอบ) ซึ่งแยกไว้สำหรับการประเมิน ในตอนท้าย เราจะรายงานผลการทำงานของโมเดลของเราบนทั้งสองส่วน คุณอาจคิดว่าประสิทธิภาพการฝึกอบรมนั้นคล้ายกับคะแนนที่นักเรียนได้รับในการสอบซ้อมที่ใช้เตรียมตัวสำหรับการสอบจริงบางอย่าง แม้ว่าผลลัพธ์จะน่าพอใจ แต่ก็ไม่ได้รับประกันความสำเร็จในการสอบจริง ในขณะที่เรียน นักเรียนอาจเริ่มท่องจำคำถามซ้อม ทำให้ดูเหมือนว่าเชี่ยวชาญในหัวข้อนั้น แต่ล้มเหลวเมื่อเผชิญกับคำถามที่ไม่เคยเห็นมาก่อนในการสอบจริง เมื่อโมเดลทำงานได้ดีกับชุดฝึกอบรมแต่ไม่สามารถทั่วไปไปเป็นข้อมูลที่ไม่เคยเห็นมาก่อน เราจะกล่าวว่าโมเดลนั้นเกิดการโอเวอร์ฟิตติ้ง (overfitting) กับชุดฝึกอบรม

**สรุป**

ฟังก์ชันวัตถุประสงค์เป็นวิธีการวัดประสิทธิภาพของโมเดลการเรียนรู้ของเครื่อง โดยทั่วไปแล้ว เราจะกำหนดฟังก์ชันวัตถุประสงค์ให้ต่ำเท่าไหร่ยิ่งดี ในการฝึกโมเดลการเรียนรู้ของเครื่อง เราจะลดการสูญเสียที่เกิดขึ้นบนชุดข้อมูลฝึกอบรม อย่างไรก็ตาม เราจำเป็นต้องระวังการโอเวอร์

1.2.4 Optimization Algorithms



เมื่อเรามีแหล่งที่มาของข้อมูล การแสดงข้อมูล โมเดล และฟังก์ชันวัตถุประสงค์ที่กำหนดไว้อย่างดีแล้ว เราจำเป็นต้องมีอัลกอริทึมที่สามารถค้นหาพารามิเตอร์ที่ดีที่สุดสำหรับการลดฟังก์ชันการสูญเสีย อัลกอริทึมการหาค่าที่ดีที่สุดที่เป็นที่นิยมสำหรับการเรียนรู้เชิงลึกนั้นอิงตามวิธีการที่เรียกว่าการไล่ระดับเชิงลาด (gradient descent) โดยสรุป ที่แต่ละขั้นตอน วิธีการนี้จะตรวจสอบเพื่อดูว่าสำหรับแต่ละพารามิเตอร์ การสูญเสียชุดฝึกอบรมจะเปลี่ยนไปอย่างไรหากคุณรบกวนพารามิเตอร์นั้นเพียงเล็กน้อย จากนั้นจะอัปเดตพารามิเตอร์ในทิศทางที่ลดการสูญเสีย

**คำอธิบายเพิ่มเติม**

การไล่ระดับเชิงลาดเป็นอัลกอริทึมการหาค่าที่ดีที่สุดแบบวนซ้ำที่ใช้ในการฝึกโมเดลการเรียนรู้ของเครื่อง โดยการเริ่มต้นด้วยชุดพารามิเตอร์แบบสุ่ม อัลกอริทึมจะคำนวณการไล่ระดับเชิงลาดของฟังก์ชันการสูญเสียที่พารามิเตอร์นั้น การไล่ระดับเชิงลาดเป็นเวกเตอร์ที่ชี้ไปในทิศทางที่ฟังก์ชันการสูญเสียเพิ่มขึ้นเร็วที่สุด อัลกอริทึมจะอัปเดตพารามิเตอร์โดยย้ายไปในทิศทางตรงข้ามของการไล่ระดับเชิงลาด ซึ่งจะลดฟังก์ชันการสูญเสีย จากนั้น อัลกอริทึมจะทำซ้ำขั้นตอนนี้จนกว่าจะบรรลุเกณฑ์การหยุดทำงาน เช่น จำนวนรอบสูงสุดหรือการเปลี่ยนแปลงของฟังก์ชันการสูญเสียต่ำกว่าเกณฑ์ที่กำหนด

การไล่ระดับเชิงลาดเป็นอัลกอริทึมที่ง่ายแต่ทรงพลังสำหรับการหาค่าที่ดีที่สุดของฟังก์ชันใดๆ ไม่เพียงแต่สำหรับฟังก์ชันการสูญเสียในโมเดลการเรียนรู้ของเครื่องเท่านั้น แต่ยังใช้ในสาขาอื่นๆ เช่น วิทยาศาสตร์ข้อมูล การเงิน และการวิจัยเชิงวิทยาศาสตร์

# 1.3 Kinds of Machine Learning Problems


ปัญหาคำปลุกในตัวอย่างการสร้างแรงจูงใจของเราเป็นเพียงหนึ่งในหลายๆ ปัญหาที่การเรียนรู้ของเครื่องสามารถรับมือได้ เพื่อสร้างแรงจูงใจให้ผู้อ่านต่อไปและให้เรามีภาษาที่ใช้ร่วมกันซึ่งจะติดตามเราไปตลอดทั้งเล่ม ตอนนี้เราจะให้ภาพรวมกว้างๆ ของภูมิทัศน์ของปัญหาด้านการเรียนรู้ของเครื่อง

1.3.1 Supervised Learning

ปัญหาคำปลุกในตัวอย่างการสร้างแรงจูงใจของเราเป็นเพียงหนึ่งในหลายๆ ปัญหาที่การเรียนรู้ของเครื่องสามารถรับมือได้ เพื่อสร้างแรงจูงใจให้ผู้อ่านต่อไปและให้เรามีภาษาที่ใช้ร่วมกันซึ่งจะติดตามเราไปตลอดทั้งเล่ม ตอนนี้เราจะให้ภาพรวมกว้างๆ ของภูมิทัศน์ของปัญหาด้านการเรียนรู้ของเครื่อง

การเรียนรู้แบบมีผู้สอน (Supervised learning) คือการเรียนรู้ประเภทหนึ่งของการเรียนรู้ของเครื่องที่โมเดลการเรียนรู้ของเครื่องถูกฝึกฝนโดยใช้ชุดข้อมูลที่มีทั้งข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว โมเดลจะเรียนรู้ที่จะทำนายข้อมูลออกจากข้อมูลเข้า

การเรียนรู้แบบมีผู้สอนมีประโยชน์มากมาย และใช้กันอย่างแพร่หลายในอุตสาหกรรม ตัวอย่างของปัญหาการเรียนรู้แบบมีผู้สอน ได้แก่

* การจำแนกภาพ: การฝึกโมเดลการเรียนรู้ของเครื่องเพื่อจำแนกภาพ เช่น การจำแนกภาพแมวหรือสุนัข
* การแปลภาษา: การฝึกโมเดลการเรียนรู้ของเครื่องเพื่อแปลภาษา เช่น การแปลประโยคภาษาอังกฤษเป็นภาษาฝรั่งเศส
* การคาดการณ์ราคาหุ้น: การฝึกโมเดลการเรียนรู้ของเครื่องเพื่อคาดการณ์ราคาหุ้นในอนาคต

กระบวนการเรียนรู้แบบมีผู้สอนมีขั้นตอนหลักๆ ดังนี้

1. รวบรวมชุดข้อมูลฝึกอบรม (training set) ซึ่งประกอบด้วยข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว
2. เลือกอัลกอริทึมการเรียนรู้ของเครื่องที่เหมาะสม
3. ฝึกโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลฝึกอบรม
4. ประเมินประสิทธิภาพของโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลทดสอบ (test set)
5. นำโมเดลการเรียนรู้ของเครื่องไปใช้งานกับข้อมูลใหม่

การเรียนรู้แบบมีผู้สอนเป็นเทคนิคการเรียนรู้ของเครื่องที่ทรงพลังและมีประโยชน์มากมาย อย่างไรก็ตาม สิ่งสำคัญคือต้องจำไว้ว่าการเรียนรู้แบบมีผู้สอนต้องใช้ชุดข้อมูลฝึกอบรมที่มีคุณภาพดี และอาจต้องใช้เวลาและทรัพยากรในการฝึกโมเดลการเรียนรู้ของเครื่อง

Regression


การถดถอย (Regression) เป็นประเภทหนึ่งของการเรียนรู้แบบมีผู้สอนที่โมเดลการเรียนรู้ของเครื่องถูกฝึกฝนโดยใช้ชุดข้อมูลที่มีทั้งข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว โดยข้อมูลออกเป็นค่าตัวเลขต่อเนื่อง เช่น ราคาบ้าน ระยะเวลาในการผ่าตัด หรือปริมาณน้ำฝน

ตัวอย่างของปัญหาการถดถอย ได้แก่

* การคาดการณ์ราคาบ้านโดยใช้ข้อมูลเข้า เช่น ขนาดพื้นที่ จำนวนห้องนอน จำนวนห้องน้ำ และระยะทางเดินเท้าไปใจกลางเมือง
* การคาดการณ์ระยะเวลาในการผ่าตัดโดยใช้ข้อมูลเข้า เช่น อายุของผู้ป่วย ประเภทของการผ่าตัด และความซับซ้อนของการผ่าตัด
* การคาดการณ์ปริมาณน้ำฝนโดยใช้ข้อมูลเข้า เช่น ความชื้น อุณหภูมิ และความเร็วลม

กระบวนการเรียนรู้การถดถอยมีขั้นตอนหลักๆ ดังนี้

1. รวบรวมชุดข้อมูลฝึกอบรม (training set) ซึ่งประกอบด้วยข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว
2. เลือกอัลกอริทึมการเรียนรู้ของเครื่องที่เหมาะสม
3. ฝึกโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลฝึกอบรม
4. ประเมินประสิทธิภาพของโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลทดสอบ (test set)
5. นำโมเดลการเรียนรู้ของเครื่องไปใช้งานกับข้อมูลใหม่

โมเดลการถดถอยที่พบบ่อย ได้แก่

* Linear regression: โมเดลการถดถอยแบบเส้นตรง เป็นโมเดลการถดถอยที่ง่ายที่สุด โดยโมเดลจะเรียนรู้ที่จะทำนายข้อมูลออกจากข้อมูลเข้าโดยใช้สมการเส้นตรง
* Logistic regression: โมเดลการถดถอยแบบลอจิสติก เป็นโมเดลการถดถอยที่ใช้ในการจำแนกประเภทแบบไบนารี โดยโมเดลจะเรียนรู้ที่จะทำนายว่าข้อมูลเข้าเป็นประเภทใดจากประเภทที่มีอยู่สองประเภท
* Polynomial regression: โมเดลการถดถอยแบบพหุนาม เป็นโมเดลการถดถอยที่ใช้ในการทำนายข้อมูลออกจากข้อมูลเข้าโดยใช้สมการพหุนาม
* Tree-based regression: โมเดลการถดถอยแบบต้นไม้ เป็นโมเดลการถดถอยที่ใช้ในการทำนายข้อมูลออกจากข้อมูลเข้าโดยใช้ต้นไม้ตัดสินใจ

การถดถอยเป็นเทคนิคการเรียนรู้ของเครื่องที่ทรงพลังและมีประโยชน์มากมาย สามารถนำไปใช้ในการแก้ปัญหาที่หลากหลายในหลากหลายอุตสาหกรรม

Classification

การจำแนกประเภท (Classification) เป็นประเภทหนึ่งของการเรียนรู้แบบมีผู้สอนที่โมเดลการเรียนรู้ของเครื่องถูกฝึกฝนโดยใช้ชุดข้อมูลที่มีทั้งข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว โดยข้อมูลออกเป็นค่าตัวเลขแบบไม่ต่อเนื่อง เช่น ชนิดของดอกไม้ ประเภทของสัตว์ หรือประเภทของโรค

ตัวอย่างของปัญหาการจำแนกประเภท ได้แก่

* การจำแนกประเภทภาพดอกไม้โดยใช้ข้อมูลเข้า เช่น สีของดอกไม้ รูปทรงของดอกไม้ และขนาดของดอกไม้
* การจำแนกประเภทภาพสัตว์โดยใช้ข้อมูลเข้า เช่น สีของขนสัตว์ รูปร่างของสัตว์ และขนาดของสัตว์
* การจำแนกประเภทโรคโดยใช้ข้อมูลเข้า เช่น อาการของผู้ป่วย ผลลัพธ์จากการตรวจเลือด และผลลัพธ์จากการตรวจภาพเอกซเรย์

กระบวนการเรียนรู้การจำแนกประเภทมีขั้นตอนหลักๆ ดังนี้

1. รวบรวมชุดข้อมูลฝึกอบรม (training set) ซึ่งประกอบด้วยข้อมูลเข้าและข้อมูลออกที่ทราบแล้ว
2. เลือกอัลกอริทึมการเรียนรู้ของเครื่องที่เหมาะสม
3. ฝึกโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลฝึกอบรม
4. ประเมินประสิทธิภาพของโมเดลการเรียนรู้ของเครื่องโดยใช้ชุดข้อมูลทดสอบ (test set)
5. นำโมเดลการเรียนรู้ของเครื่องไปใช้งานกับข้อมูลใหม่

โมเดลการจำแนกประเภทที่พบบ่อย ได้แก่

* Logistic regression: โมเดลการจำแนกประเภทแบบลอจิสติก เป็นโมเดลการจำแนกประเภทที่ใช้ในการจำแนกประเภทแบบไบนารี โดยโมเดลจะเรียนรู้ที่จะทำนายว่าข้อมูลเข้าเป็นประเภทใดจากประเภทที่มีอยู่สองประเภท
* Support vector machine (SVM): โมเดลการจำแนกประเภทแบบ SVM เป็นโมเดลการจำแนกประเภทที่ใช้ในการจำแนกประเภททั้งแบบไบนารีและแบบมัลติคลาส โดยโมเดลจะเรียนรู้ที่จะสร้างเส้นไฮเปอร์เพลน (hyperplane) เพื่อแบ่งข้อมูลออกเป็นประเภทต่างๆ
* Decision tree: โมเดลการจำแนกประเภทแบบต้นไม้ตัดสินใจ เป็นโมเดลการจำแนกประเภทที่ใช้ในการจำแนกประเภททั้งแบบไบนารีและแบบมัลติคลาส โดยโมเดลจะเรียนรู้ที่จะสร้างต้นไม้ตัดสินใจเพื่อจำแนกประเภทข้อมูล

การจำแนกประเภทเป็นเทคนิคการเรียนรู้ของเครื่องที่ทรงพลังและมีประโยชน์มากมาย สามารถนำไปใช้ในการแก้ปัญหาที่หลากหลายในหลากหลายอุตสาหกรรม เช่น การตลาด การโฆษณา การธนาคาร การแพทย์ และความมั่นคงปลอดภัย

**การแปล**

ความน่าจะเป็นที่ทำนายว่าเห็ดในรูปที่ 1.3.2 เป็นเห็ดพิษอาจเท่ากับ 0.2 × ∞ + 0.8 × 0 = ∞ ในขณะที่การสูญเสียจากการทิ้งเห็ดนั้นคือ 0.2 × 0 + 0.8 × 1 = 0.8 ดังนั้นความระมัดระวังของเราจึงเป็นสิ่งที่ถูกต้อง เพราะเห็ดในรูปที่ 1.3.2 นั้นเป็นเห็ดพิษจริงๆ

การจำแนกประเภทอาจมีความซับซ้อนมากกว่าการจำแนกประเภทแบบไบนารีหรือมัลติคลาส เช่น มีการจำแนกประเภทบางประเภทที่ใช้กับคลาสที่มีโครงสร้างแบบลำดับชั้น ในกรณีเช่นนี้ ข้อผิดพลาดไม่ใช่ทั้งหมดที่เท่ากัน หากเราต้องทำผิดพลาด เราอาจต้องการจำแนกผิดเป็นคลาสที่เกี่ยวข้องมากกว่าคลาสที่ห่างไกล ซึ่งโดยปกติแล้วเรียกว่าการจำแนกประเภทแบบลำดับชั้น ตัวอย่างเช่น Linnaeus ซึ่งจัดประเภทสัตว์เป็นลำดับชั้น

ในการจำแนกประเภทสัตว์ การจำแนกผิดพลาดระหว่างพุดเดิ้ลกับชเนาเซอร์อาจไม่ใช่เรื่องเลวร้ายนัก แต่โมเดลของเราจะต้องเสียค่าปรับอย่างมากหากจำแนกผิดพลาดระหว่างพุดเดิ้ลกับไดโนเสาร์ ลำดับชั้นใดที่เกี่ยวข้องอาจขึ้นอยู่กับวิธีที่คุณวางแผนจะใช้โมเดล ตัวอย่างเช่น งูหางกระดิ่งและงูรัดเกิร์ตอาจใกล้เคียงกันในต้นไม้วิวัฒนาการ แต่การจำแนกผิดพลาดระหว่างงูหางกระดิ่งกับงูรัดเกิร์ตอาจมีผลลัพธ์ที่ร้ายแรง

**หมายเหตุ**

* Linnaeus เป็นนักพฤกษศาสตร์และสัตวแพทย์ชาวสวีเดนที่เป็นผู้บัญญัติระบบการแบ่งประเภทสิ่งมีชีวิตแบบลำดับชั้นสมัยใหม่
* การจัดประเภทสัตว์ตามลำดับชั้นของ Linnaeus มีอยู่ 7 ระดับ ได้แก่ ราชอาณาจักร (kingdom), ไฟลัม (phylum), ชั้น (class), อันดับ (order), วงศ์ (family), สกุล (genus) และชนิด (species)
* งูหางกระดิ่งเป็นงูพิษที่มีพิษร้ายแรงที่สุดชนิดหนึ่งในโลก ส่วนงูรัดเกิร์ตเป็นงูที่ไม่มีพิษ

ปัญหาการจำแนกประเภทบางอย่างเหมาะสมกับการจำแนกประเภทแบบไบนารีหรือแบบมัลติคลาส เช่น เราสามารถฝึกตัวจำแนกประเภทแบบไบนารีตามปกติเพื่อแยกแยะแมวออกจากสุนัข ด้วยเทคโนโลยีการประมวลผลภาพในปัจจุบัน เราสามารถทำสิ่งนี้ได้อย่างง่ายดายโดยใช้เครื่องมือสำเร็จรูป อย่างไรก็ตาม ไม่ว่าโมเดลของเราจะแม่นยำแค่ไหน เราอาจพบว่าตัวเองมีปัญหาเมื่อตัวจำแนกประเภทพบกับรูปภาพของ Town Musicians of Bremen ซึ่งเป็นนิทานพื้นบ้านเยอรมันยอดนิยมที่มีสัตว์สี่ตัว (รูปที่ 1.3.3)

อย่างที่คุณเห็น ภาพนี้มีแมว ไก่ฟ้า สุนัข และลา พร้อมกับต้นไม้ที่อยู่เบื้องหลัง หากเราคาดการณ์ว่าจะพบกับภาพดังกล่าว การจำแนกประเภทแบบมัลติคลาสอาจไม่ใช่การกำหนดปัญหาที่เหมาะสม แทนที่จะเป็นเช่นนั้น เราอาจต้องการให้โมเดลมีตัวเลือกในการบอกว่ารูปภาพแสดงแมว สุนัข ลา และไก่ฟ้า

ปัญหาการเรียนรู้ที่จะทำนายคลาสที่ไม่แยกกันเป็นอิสระจากกันนั้นเรียกว่าการจำแนกประเภทแบบมัลติเลเบล โดยปกติแล้ว ปัญหาการแท็กอัตโนมัติจะอธิบายได้ดีที่สุดในแง่ของการจำแนกประเภทแบบมัลติเลเบล ลองนึกถึงแท็กที่ผู้คนอาจนำไปใช้กับโพสต์บนบล็อกด้านเทคนิค เช่น "การเรียนรู้ของเครื่อง", "เทคโนโลยี", "แกดเจ็ต", "ภาษาการเขียนโปรแกรม", "Linux", "การประมวลผลคลาวด์", "AWS" บทความทั่วไปอาจมีแท็ก 5–10 แท็ก โดยปกติ แท็กจะมีความสัมพันธ์กันบ้าง โพสต์เกี่ยวกับ "การประมวลผลคลาวด์" มักจะกล่าวถึง "AWS" และโพสต์เกี่ยวกับ "การเรียนรู้ของเครื่อง" มักจะกล่าวถึง "GPUs"

การจัดอันดับกลุ่มรายการเป็นสิ่งที่พบได้บ่อยในด้านการค้นหาข้อมูล ตัวอย่างเช่น การค้นหาบนเว็บเป้าหมายไม่ใช่การพิจารณาว่าหน้าเว็บใดหน้าหนึ่งเกี่ยวข้องกับคำค้นหาหรือไม่ แต่เป็นการพิจารณาว่าหน้าเว็บใดในกลุ่มผลลัพธ์ที่เกี่ยวข้อง ควรแสดงให้เห็นอย่างเด่นชัดที่สุดแก่ผู้ใช้รายใดรายหนึ่ง วิธีหนึ่งในการทำเช่นนี้คือการกำหนดคะแนนให้กับแต่ละองค์ประกอบในกลุ่มก่อน จากนั้นค้นหาองค์ประกอบที่มีคะแนนสูงสุด PageRank ซึ่งเป็นสูตรลับเบื้องหลังเครื่องมือค้นหา Google ในยุคแรกเป็นตัวอย่างแรกของระบบการให้คะแนนประเภทนี้

แปลกที่คะแนนที่ PageRank ให้ไม่ได้ขึ้นอยู่กับคำค้นหาจริง แทนที่จะเป็นเช่นนั้น พวกเขาใช้ตัวกรองความเกี่ยวข้องอย่างง่ายๆ เพื่อระบุชุดของผู้สมัครที่เกี่ยวข้อง จากนั้นใช้ PageRank เพื่อจัดลำดับความสำคัญของหน้าเว็บที่มีอำนาจมากขึ้น ปัจจุบัน เครื่องมือค้นหาใช้การเรียนรู้ของเครื่องและโมเดลพฤติกรรมเพื่อรับคะแนนความเกี่ยวข้องที่ขึ้นอยู่กับคำค้นหา มีการประชุมวิชาการทั้งหมดที่อุทิศให้กับหัวข้อนี้


Recommender System

**ระบบแนะนำ** (Recommender system) เป็นอีกหนึ่งปัญหาที่เกี่ยวข้องกับการค้นหาและการจัดอันดับ ปัญหามีความคล้ายคลึงกันในแง่ที่เป้าหมายคือการแสดงชุดของรายการที่เกี่ยวข้องกับผู้ใช้ ความแตกต่างที่สำคัญคือการเน้นย้ำถึงการปรับแต่งให้เหมาะสมกับผู้ใช้เฉพาะในบริบทของระบบแนะนำ ตัวอย่างเช่น สำหรับคำแนะนำภาพยนตร์ ผลลัพธ์สำหรับแฟนไซไฟและผลลัพธ์สำหรับผู้เชี่ยวชาญด้านภาพยนตร์ตลกของ Peter Sellers อาจแตกต่างกันอย่างมาก ปัญหาที่คล้ายคลึงกันปรากฏขึ้นในการตั้งค่าคำแนะนำอื่นๆ เช่น สำหรับผลิตภัณฑ์ขายปลีก เพลง และคำแนะนำข่าว

ในบางกรณี ลูกค้าให้ข้อเสนอแนะอย่างชัดเจน โดยแจ้งให้ทราบว่าพวกเขาชอบผลิตภัณฑ์ใดผลิตภัณฑ์หนึ่งมากแค่ไหน (เช่น คะแนนและรีวิวผลิตภัณฑ์บน Amazon, IMDb หรือ Goodreads) ในกรณีอื่นๆ พวกเขาให้ข้อเสนอแนะโดยนัย เช่น โดยข้ามเพลงในเพลย์ลิสต์ ซึ่งอาจบ่งบอกถึงความไม่พึงพอใจหรืออาจบ่งบอกว่าเพลงนั้นไม่เหมาะสมในบริบท ในสูตรที่ง่ายที่สุด ระบบเหล่านี้ได้รับการฝึกฝนให้ประมาณคะแนน เช่น คะแนนดาวที่คาดหวังหรือความน่าจะเป็นที่ผู้ใช้ที่กำหนดจะซื้อไอเท็มใดไอเท็มหนึ่ง

จากแบบจำลองดังกล่าว สำหรับผู้ใช้ที่กำหนด เราสามารถเรียกค้นชุดของวัตถุที่มีคะแนนสูงสุด ซึ่งจากนั้นสามารถแนะนำให้กับผู้ใช้ได้ ระบบการผลิตมีความก้าวหน้ากว่ามากและพิจารณาถึงกิจกรรมของผู้ใช้และลักษณะของสินค้าอย่างละเอียดเมื่อคำนวณคะแนนดังกล่าว รูปที่ 1.3.4 แสดงหนังสือการเรียนรู้เชิงลึกที่ Amazon แนะนำตามอัลกอริทึมการปรับแต่งให้เหมาะสมกับความชอบของ Aston

แม้ว่าระบบแนะนำที่สร้างขึ้นบนโมเดลการทำนายจะมีมูลค่าทางเศรษฐกิจมหาศาล แต่ก็มีข้อบกพร่องเชิงแนวคิดที่ร้ายแรงบางประการ ประการแรก เราสังเกตเห็นข้อเสนอแนะที่ผ่านการเซ็นเซอร์เท่านั้น: ผู้ใช้ชอบให้คะแนนภาพยนตร์ที่รู้สึกว่าชอบจริงๆ ตัวอย่างเช่น ในมาตราส่วนห้าจุด คุณอาจสังเกตเห็นว่ารายการได้รับคะแนนหนึ่งและห้าดาวมากมาย แต่มีคะแนนสามดาวน้อยอย่างเห็นได้ชัด นอกจากนี้ พฤติกรรมการซื้อปัจจุบันมักเป็นผลมาจากอัลกอริทึมการแนะนำที่มีอยู่ในปัจจุบัน แต่ขั้นตอนการเรียนรู้ไม่ได้คำนึงถึงรายละเอียดนี้เสมอไป ดังนั้น จึงเป็นไปได้ที่ห่วงโซ่ข้อเสนอแนะจะเกิดขึ้นเมื่อระบบแนะนำผลักดันรายการที่ผู้ใช้ชอบเป็นพิเศษ จากนั้นจึงถือว่าดีกว่า (เนื่องจากการซื้อมากขึ้น) และในทางกลับกันก็ได้รับการแนะนำมากขึ้น

ปัญหาเหล่านี้หลายประการ เช่น วิธีจัดการกับการเซ็นเซอร์ สิ่งจูงใจ และห่วงโซ่ข้อเสนอแนะ เป็นคำถามการวิจัยที่เปิดกว้างและสำคัญ

Sequence Learning


การเรียนรู้แบบลำดับ (Sequence learning) เป็นประเภทของการเรียนรู้ของเครื่องที่เกี่ยวข้องกับปัญหาที่ข้อมูลเข้าและข้อมูลออกเป็นลำดับที่มีความยาวแตกต่างกัน ปัญหาเหล่านี้ต้องใช้โมเดลที่จะประมวลผลลำดับของข้อมูลเข้าหรือสร้างลำดับของข้อมูลออก (หรือทั้งสองอย่าง)

ตัวอย่างของปัญหาการเรียนรู้แบบลำดับ ได้แก่

* การแปลภาษาโดยเครื่อง (machine translation): การแปลประโยคจากภาษาหนึ่งไปเป็นอีกภาษาหนึ่ง
* การถอดเสียงพูดเป็นข้อความ (speech-to-text transcription): การแปลงคำพูดเป็นข้อความ
* การติดแท็กข้อความ (text tagging): การกำหนดแท็กให้กับคำแต่ละคำในข้อความ เช่น แท็กประเภทของคำ (part-of-speech tagging) และแท็กชื่อเฉพาะ (named entity tagging)
* การวิเคราะห์ประโยค (parsing): การวิเคราะห์โครงสร้างของประโยค

การเรียนรู้แบบลำดับเป็นหนึ่งในประยุกต์ใช้งานการเรียนรู้ของเครื่องที่น่าตื่นเต้นที่สุด และเป็นพื้นฐานของโมเดลที่ประสบความสำเร็จมากมาย เช่น Google Translate และ Siri

การติดแท็กข้อความและการวิเคราะห์ประโยคเป็นปัญหาการเรียนรู้แบบลำดับประเภทหนึ่งที่ข้อมูลเข้าและข้อมูลออกเป็นลำดับของคำที่สอดคล้องกัน ตัวอย่างเช่น ในการติดแท็กประเภทของคำ เราจะกำหนดแท็กประเภทของคำให้กับคำแต่ละคำในประโยค เช่น "noun" หรือ "direct object" ในการวิเคราะห์ประโยค เราจะวิเคราะห์โครงสร้างของประโยคเพื่อหาความสัมพันธ์ระหว่างคำต่างๆ เช่น ระหว่างประธานกับกรรม

ตัวอย่างการติดแท็กประเภทของคำ:

```
ประโยค: สุนัขกินข้าว
แท็กประเภทของคำ: N V N
```

ตัวอย่างการวิเคราะห์ประโยค:

```
ประโยค: สุนัขกินข้าว
วิเคราะห์โครงสร้าง: S V O
```

การเรียนรู้แบบลำดับเป็นหัวข้อที่กว้างขวางและซับซ้อน แต่เป็นหัวข้อที่สำคัญมากสำหรับการเรียนรู้ของเครื่อง เนื่องจากปัญหาการเรียนรู้แบบลำดับพบได้ทั่วไปในโลกจริง


การจดจำเสียงพูดอัตโนมัติ (Automatic Speech Recognition) เป็นปัญหาการเรียนรู้แบบลำดับประเภทหนึ่งที่ข้อมูลเข้าเป็นบันทึกเสียงของผู้พูด และข้อมูลออกเป็นคำถอดเสียงของสิ่งที่ผู้พูดพูด ความท้าทายคือมีเฟรมเสียง (sound frame) มากกว่าข้อความมากมาย (เสียงมักจะถูกสุ่มตัวอย่างที่ 8kHz หรือ 16kHz) นั่นคือไม่มีความสัมพันธ์ 1:1 ระหว่างเสียงกับข้อความ เนื่องจากอาจมีแซมเปิลหลายพันตัวที่สอดคล้องกับคำพูดเพียงคำเดียว ปัญหาเหล่านี้เป็นปัญหาการเรียนรู้แบบลำดับต่อลำดับที่ข้อมูลออกมีขนาดเล็กกว่าข้อมูลเข้ามาก ในขณะที่มนุษย์มีความสามารถในการจดจำเสียงพูดได้อย่างน่าทึ่ง แม้กระทั่งจากเสียงคุณภาพต่ำ แต่การทำให้คอมพิวเตอร์ทำสิ่งเดียวกันนั้นเป็นความท้าทายที่ยิ่งใหญ่

การแปลงข้อความเป็นคำพูด (Text to Speech) เป็นกระบวนการผกผันของการจดจำเสียงพูดอัตโนมัติ ในกรณีนี้ ข้อมูลเข้าคือข้อความและข้อมูลออกคือไฟล์เสียง ในกรณีนี้ ข้อมูลออกมีขนาดใหญ่กว่าข้อมูลเข้ามาก

การแปลภาษาโดยเครื่อง (Machine Translation) แตกต่างจากกรณีของการจดจำเสียงพูดที่ข้อมูลเข้าและข้อมูลออกที่สอดคล้องกันเกิดขึ้นในลำดับเดียวกัน ในการแปลภาษาโดยเครื่อง ข้อมูลที่ไม่ได้จัดเรียงกันนั้นเป็นความท้าทายใหม่ ในกรณีนี้ ลำดับข้อมูลเข้าและข้อมูลออกอาจมีความยาวต่างกัน และส่วนที่สอดคล้องกันของลำดับที่เกี่ยวข้องอาจปรากฏในลำดับที่ต่างกัน พิจารณาตัวอย่างที่ชัดเจนต่อไปนี้เกี่ยวกับแนวโน้มที่แปลกประหลาดของชาวเยอรมันในการวางกริยาไว้ที่ท้ายประโยค:

```
ภาษาเยอรมัน: Ich habe ein Buch gelesen.
ภาษาอังกฤษ: I have read a book.
```

ปัญหาที่เกี่ยวข้องจำนวนมากพบได้ในงานการเรียนรู้อื่นๆ ตัวอย่างเช่น การกำหนดลำดับที่ผู้ใช้อ่านเว็บเพจเป็นปัญหาการวิเคราะห์เค้าโครงสองมิติ ปัญหาการโต้ตอบแสดงความซับซ้อนเพิ่มเติมทุกประเภท ซึ่งการกำหนดว่าจะพูดอะไรต่อไปจำเป็นต้องคำนึงถึงความรู้ในโลกแห่งความจริงและสถานะก่อนหน้าของการสนทนาในระยะเวลาที่ยาวนาน หัวข้อเหล่านี้เป็นหัวข้อวิจัยที่สำคัญ

1.3.2 Unsupervised and Self-Supervised Learning


ตัวอย่างก่อนหน้านี้มุ่งเน้นไปที่การเรียนรู้แบบมีผู้สอน (supervised learning) ซึ่งเราป้อนข้อมูลขนาดใหญ่ที่มีทั้งคุณลักษณะและค่าป้ายกำกับที่สอดคล้องกันให้กับโมเดล คุณสามารถคิดว่าผู้เรียนแบบมีผู้สอนมีงานที่เฉพาะเจาะจงมากและมีเจ้านายที่เข้มงวดมาก เจ้านายจะคอยยืนอยู่ข้างหลังผู้เรียนและบอกพวกเขาว่าต้องทำอะไรในทุกสถานการณ์จนกว่าพวกเขาจะเรียนรู้ที่จะจับคู่จากสถานการณ์ไปจนถึงการกระทำ การทำงานให้กับเจ้านายแบบนี้ดูน่าเบื่อมาก อย่างไรก็ตาม การทำให้เจ้านายแบบนี้พอใจนั้นค่อนข้างง่าย คุณเพียงแค่จดจำรูปแบบให้เร็วที่สุดเท่าที่จะทำได้และเลียนแบบการกระทำของเจ้านาย

เมื่อพิจารณาถึงสถานการณ์ตรงกันข้าม การทำงานให้กับเจ้านายที่ไม่รู้ว่าพวกเขาต้องการให้คุณทำอะไรนั้นอาจน่าหงุดหงิด อย่างไรก็ตาม หากคุณวางแผนที่จะเป็นนักวิทยาศาสตร์ข้อมูล คุณจะต้องชินกับมัน เจ้านายอาจมอบข้อมูลขนาดใหญ่ให้คุณและบอกให้คุณทำวิทยาศาสตร์ข้อมูลกับข้อมูลนั้น! ฟังดูคลุมเครือเพราะมันคลุมเครือ เราเรียกปัญหาประเภทนี้ว่าการเรียนรู้แบบไม่มีผู้สอน (unsupervised learning) และประเภทและจำนวนคำถามที่เราสามารถถามได้นั้นถูกจำกัดด้วยความคิดสร้างสรรค์ของเราเท่านั้น เราจะกล่าวถึงเทคนิคการเรียนรู้แบบไม่มีผู้สอนในบทถัดไป เพื่อเรียกน้ำย่อยของคุณในตอนนี้ เราจะอธิบายคำถามบางส่วนต่อไปนี้ที่คุณอาจถาม

* เราสามารถหาต้นแบบจำนวนน้อยที่สรุปข้อมูลได้อย่างถูกต้องหรือไม่? จากภาพถ่ายชุดหนึ่ง เราสามารถจัดกลุ่มรูปภาพเป็นภาพถ่ายทิวทัศน์ รูปภาพสุนัข ทารก แมว และยอดเขาได้หรือไม่? ในทำนองเดียวกัน จากการรวบรวมกิจกรรมการท่องเว็บของผู้ใช้ เราสามารถจัดกลุ่มผู้ใช้ที่มีพฤติกรรมคล้ายกันได้หรือไม่? ปัญหานี้มักเรียกว่าการจัดกลุ่ม (clustering)
* เราสามารถหาพารามิเตอร์จำนวนน้อยที่สามารถจับคุณสมบัติที่เกี่ยวข้องของข้อมูลได้อย่างถูกต้องหรือไม่? เส้นทางการเคลื่อนที่ของลูกบอลสามารถอธิบายได้ด้วยความเร็ว เส้นผ่านศูนย์กลาง และมวลของลูกบอล ช่างตัดเย็บได้พัฒนาพารามิเตอร์จำนวนเล็กน้อยที่อธิบายรูปร่างร่างกายมนุษย์อย่างแม่นยำพอสมควรเพื่อจุดประสงค์ในการตัดเย็บเสื้อผ้า ปัญหาเหล่านี้เรียกว่าการประมาณค่าซับสเปซ (subspace estimation) หากความสัมพันธ์เป็นเชิงเส้น จะเรียกว่าการวิเคราะห์องค์ประกอบหลัก (principal component analysis)
* มีการแสดงของวัตถุ (ที่มีโครงสร้างตามอำเภอใจ) ในพื้นที่แบบยุคลิดหรือไม่ เพื่อให้คุณสมบัติแบบสัญลักษณ์สามารถจับคู่กันได้อย่างดี? วิธีนี้ใช้เพื่ออธิบายเอนทิตีและความสัมพันธ์ของพวกมัน เช่น "Rome" - "Italy" + "France" = "Paris"
* มีคำอธิบายเกี่ยวกับสาเหตุหลักของข้อมูลจำนวนมากที่เราสังเกตเห็นหรือไม่? ตัวอย่างเช่น หากเรามีข้อมูลประชากรเกี่ยวกับราคาบ้าน มลพิษ อาชญากรรม ทำเลที่ตั้ง การศึกษา และเงินเดือน เราสามารถค้นพบว่าข้อมูลเหล่านี้เกี่ยวข้องกันอย่างไรโดยเพียงแค่พิจารณาจากข้อมูลเชิงประจักษ์? สาขาที่เกี่ยวข้องกับความสัมพันธ์เชิงเหตุผลและโมเดลกราฟิกแบบมีโอกาสจะรับมือกับคำถามประเภทนี้
* ความก้าวหน้าล่าสุดที่สำคัญและน่าตื่นเต้นอีกประการหนึ่งในการเรียนรู้แบบไม่มีผู้สอนคือการถือกำเนิดของโมเดลการสร้างแบบจำลองเชิงลึก (deep generative models) โมเดลเหล่านี้ประมาณความหนาแน่นของข้อมูลอย่างชัดเจนหรือโดยนัย หลังจากฝึกอบรมแล้ว เราสามารถใช้โมเดลการสร้างแบบจำลองเพื่อประเมินตัวอย่างตามความน่าจะเป็น หรือเพื่อสุ่มตัวอย่างตัวอย่างสังเคราะห์จากการแจกแจงที่เรียนรู้ได้ ความก้าวหน้าในยุคแรกของการเรียนรู้แบบลึกในการสร้างแบบจำลองเชิงลึกเกิดขึ้นพร้อมกับการประดิษ

1.3.3 Interacting with an Environment

จนถึงตอนนี้ เรายังไม่ได้กล่าวถึงที่มาของข้อมูลหรือสิ่งที่เกิดขึ้นจริงเมื่อโมเดลการเรียนรู้ของเครื่องสร้างเอาต์พุต เนื่องจากการเรียนรู้แบบมีผู้สอนและการเรียนรู้แบบไม่มีผู้สอนไม่ได้กล่าวถึงปัญหเหล่านี้ในลักษณะที่ซับซ้อนมาก ในแต่ละกรณี เราจะรวบรวมข้อมูลจำนวนมากล่วงหน้า จากนั้นตั้งค่าเครื่องจักรการจดจำรูปแบบของเราโดยไม่ต้องโต้ตอบกับสภาพแวดล้อมอีกต่อไป เนื่องจากการเรียนรู้ทั้งหมดเกิดขึ้นหลังจากอัลกอริทึมถูกตัดการเชื่อมต่อจากสภาพแวดล้อม สิ่งนี้จึงบางครั้งเรียกว่าการเรียนรู้แบบออฟไลน์ (offline learning) ตัวอย่างเช่น การเรียนรู้แบบมีผู้สอนถือว่ารูปแบบการโต้ตอบที่เรียบง่ายดังที่แสดงในรูปที่ 1.3.6

ความเรียบง่ายของการเรียนรู้แบบออฟไลน์นั้นมีเสน่ห์ ข้อดีคือเราสามารถกังวลเกี่ยวกับการจดจำรูปแบบโดยไม่ต้องกังวลเกี่ยวกับความยุ่งยากที่เกิดจากการโต้ตอบกับสภาพแวดล้อมแบบไดนามิก แต่การกำหนดสูตรปัญหานี้มีข้อจำกัด หากคุณเติบโตมากับการอ่านนวนิยาย Robot ของ Asimov คุณอาจจินตนาการถึงเอเย่นต์อัจฉริยะประดิษฐ์ที่ไม่เพียงแต่สามารถทำนายได้ แต่ยังสามารถดำเนินการในโลกได้อีกด้วย เราต้องการคิดเกี่ยวกับเอเย่นต์อัจฉริยะ ไม่ใช่แค่โมเดลการทำนาย ซึ่งหมายความว่าเราต้องคิดเกี่ยวกับการเลือกการกระทำ ไม่ใช่แค่การทำนาย ในทางตรงกันข้ามกับการทำนายเพียงอย่างเดียว การกระทำนั้นส่งผลกระทบต่อสภาพแวดล้อมจริง ๆ หากเราต้องการฝึกอบรมเอเย่นต์อัจฉริยะ เราต้องคิดเกี่ยวกับวิธีการโต้ตอบกับสภาพแวดล้อมในลักษณะที่มีประโยชน์

นี่คือที่ที่การเรียนรู้แบบเสริมแรง (reinforcement learning) เข้ามา การเรียนรู้แบบเสริมแรงเป็นประเภทของการเรียนรู้ของเครื่องที่เอเย่นต์เรียนรู้โดยการโต้ตอบกับสภาพแวดล้อม เอเย่นต์จะเลือกลงมือทำในสิ่งที่คิดว่าดีที่สุดในสถานการณ์ปัจจุบัน จากนั้นจะได้รับรางวัลหรือการลงโทษจากสภาพแวดล้อม จากนั้นเอเย่นต์จะปรับปรุงนโยบายการกระทำของตนเพื่อให้ได้รับรางวัลมากขึ้นและหลีกเลี่ยงการลงโทษ

การเรียนรู้แบบเสริมแรงเป็นเทคนิคที่ทรงพลังมาก ซึ่งถูกใช้เพื่อฝึกอบรมเอเย่นต์อัจฉริยะในหลากหลายแอปพลิเคชัน ตั้งแต่การเล่นเกมจนถึงการควบคุมหุ่นยนต์ การเรียนรู้แบบเสริมแรงนั้นซับซ้อนกว่าการเรียนรู้แบบมีผู้สอนและการเรียนรู้แบบไม่มีผู้สอน เนื่องจากเอเย่นต์ต้องเรียนรู้โดยการทดลองผิดพลาด แต่ก็มีความยืดหยุ่นและทรงพลังมากขึ้น

คำถามเหล่านี้ก่อให้เกิดปัญหาการเปลี่ยนแปลงการแจกแจง ซึ่งข้อมูลการฝึกอบรมและข้อมูลทดสอบแตกต่างกัน ตัวอย่างของสิ่งนี้ที่หลายคนอาจเคยพบคือ การทำข้อสอบที่เขียนโดยอาจารย์ ในขณะที่งานบ้านนั้นแต่งโดยผู้ช่วยสอน

1.3.4 Reinforcement Learning


การเรียนรู้แบบเสริมแรง (reinforcement learning) เป็นประเภทของการเรียนรู้ของเครื่องที่เอเย่นต์เรียนรู้โดยการโต้ตอบกับสภาพแวดล้อม เอเย่นต์จะเลือกลงมือทำในสิ่งที่คิดว่าดีที่สุดในสถานการณ์ปัจจุบัน จากนั้นจะได้รับรางวัลหรือการลงโทษจากสภาพแวดล้อม จากนั้นเอเย่นต์จะปรับปรุงนโยบายการกระทำของตนเพื่อให้ได้รับรางวัลมากขึ้นและหลีกเลี่ยงการลงโทษ

การเรียนรู้แบบเสริมแรงเป็นเทคนิคที่ทรงพลังมาก ซึ่งถูกใช้เพื่อฝึกอบรมเอเย่นต์อัจฉริยะในหลากหลายแอปพลิเคชัน ตั้งแต่การเล่นเกมจนถึงการควบคุมหุ่นยนต์ การเรียนรู้แบบเสริมแรงนั้นซับซ้อนกว่าการเรียนรู้แบบมีผู้สอนและการเรียนรู้แบบไม่มีผู้สอน เนื่องจากเอเย่นต์ต้องเรียนรู้โดยการทดลองผิดพลาด แต่ก็มีความยืดหยุ่นและทรงพลังมากขึ้น

การเรียนรู้แบบเสริมแรงให้คำจำกัดความที่เป็นทั่วไปมากของปัญหาที่เอเย่นต์โต้ตอบกับสภาพแวดล้อมในช่วงเวลาหลายขั้น ที่แต่ละขั้นเวลา เอเย่นต์จะได้รับการสังเกตบางอย่างจากสภาพแวดล้อมและต้องเลือกลงมือทำที่ส่งต่อไปยังสภาพแวดล้อมโดยใช้กลไกบางอย่าง (บางครั้งเรียกว่าแอคชูเอเตอร์) จากนั้นหลังจากแต่ละลูป เอเย่นต์จะได้รับรางวัลจากสภาพแวดล้อม กระบวนการนี้แสดงในรูปที่ 1.3.7 จากนั้นเอเย่นต์จะได้รับการสังเกตครั้งต่อไป และเลือกลงมือทำครั้งต่อไป และอื่นๆ พฤติกรรมของเอเย่นต์การเรียนรู้แบบเสริมแรงถูกควบคุมโดยนโยบาย นโยบายสั้นๆ ก็คือฟังก์ชันที่แมปจากการสังเกตสภาพแวดล้อมไปยังการกระทำ เป้าหมายของการเรียนรู้แบบเสริมแรงคือการสร้างนโยบายที่ดี

การเรียนรู้แบบเสริมแรงเชิงลึก (deep reinforcement learning) ซึ่งนำการเรียนรู้เชิงลึกมาใช้กับปัญหาการเรียนรู้แบบเสริมแรง ได้รับความนิยมอย่างมาก ตัวอย่างที่สำคัญสองประการคือ เครือข่าย Q เชิงลึกที่ล้ำสมัย ซึ่งเอาชนะมนุษย์ที่เกม Atari โดยใช้เพียงอินพุตภาพ (Mnih et al., 2015) และโปรแกรม AlphaGo ซึ่งโค่นแชมป์โลกที่เกมกระดาน Go (Silver et al., 2016)

การเรียนรู้แบบเสริมแรงเป็นเทคนิคที่ทรงพลังมากสำหรับการฝึกอบรมเอเย่นต์อัจฉริยะให้โต้ตอบกับสภาพแวดล้อมและดำเนินการ ซึ่งมีประโยชน์สำหรับหลากหลายแอปพลิเคชัน

การเรียนรู้แบบเสริมแรงแบบทั่วไปนั้นมีความซับซ้อนสูง ซึ่งการคำนึงถึงความซับซ้อนทั้งหมดนี้พร้อมกันอาจเป็นเรื่องที่มากเกินไป นอกจากนี้ ปัญหาในทางปฏิบัติทุกประการไม่ได้แสดงความซับซ้อนทั้งหมดนี้ ดังนั้น นักวิจัยจึงได้ศึกษาปัญหาการเรียนรู้แบบเสริมแรงในกรณีพิเศษจำนวนหนึ่ง

เมื่อสภาพแวดล้อมถูกสังเกตอย่างสมบูรณ์ เราเรียกปัญหาด้านการเรียนรู้แบบเสริมแรงว่ากระบวนการตัดสินใจแบบมาร์คอฟ (Markov decision process) เมื่อสถานะไม่ขึ้นกับการกระทำก่อนหน้านี้ เราจะเรียกมันว่าปัญหาการปล้นโจรสลัดตามบริบท (contextual bandit problem) เมื่อไม่มีสถานะ มีเพียงชุดของการกระทำที่มีอยู่กับรางวัลที่ไม่ทราบแน่ชัดในตอนแรก เรามีปัญหาการปล้นโจรสลัดแบบหลายแขนแบบคลาสสิก (multi-armed bandit problem)

* กระบวนการตัดสินใจแบบมาร์คอฟ (Markov decision process) เป็นแบบจำลองคณิตศาสตร์ที่ใช้ในการแก้ไขปัญหาการควบคุมแบบวนซ้ำ ในแต่ละขั้นตอนของกระบวนการ ตัวแทนจะอยู่ในสถานะบางอย่างและสามารถเลือกการกระทำใดๆ จากชุดของการกระทำที่เป็นไปได้ เมื่อเลือกการกระทำแล้ว ตัวแทนจะได้รับรางวัลบางอย่างและย้ายไปสู่สถานะใหม่ สถานะใหม่ขึ้นอยู่กับสถานะปัจจุบันและการกระทำที่เลือก

* ปัญหาการปล้นโจรสลัดตามบริบท (contextual bandit problem) เป็นปัญหาการเรียนรู้เครื่องที่ตัวแทนต้องเลือกการกระทำที่เหมาะสมจากชุดของการกระทำที่มีอยู่โดยได้รับข้อมูลบริบทบางอย่าง เป้าหมายคือการเลือกการกระทำที่ให้รางวัลสูงสุดโดยเฉลี่ยเมื่อเวลาผ่านไป

* ปัญหาการปล้นโจรสลัดแบบหลายแขนแบบคลาสสิก (multi-armed bandit problem) เป็นปัญหาการเรียนรู้เครื่องที่ตัวแทนต้องเลือกแขนที่เหมาะสมจากชุดของแขนที่มีอยู่โดยไม่ทราบความน่าจะเป็นของรางวัลสำหรับแต่ละแขน เป้าหมายคือการเลือกแขนที่ให้รางวัลสูงสุดโดยเฉลี่ยเมื่อเวลาผ่านไป

กรณีพิเศษเหล่านี้มีความซับซ้อนน้อยกว่าปัญหาการเรียนรู้แบบเสริมแรงทั่วไป แต่ยังคงเป็นปัญหาที่ท้าทายในการแก้ไข นักวิจัยได้พัฒนาอ算法ที่หลากหลายสำหรับการแก้ไขปัญหาการเรียนรู้แบบเสริมแรงประเภทต่างๆ เหล่านี้

#1.4 Roots



เราเพิ่งทบทวนปัญหาชุดเล็กๆ ที่การเรียนรู้ของเครื่องสามารถแก้ไขได้ สำหรับปัญหาการเรียนรู้ของเครื่องที่หลากหลาย การเรียนรู้เชิงลึกให้เครื่องมือที่มีประสิทธิภาพสำหรับการแก้ปัญหา แม้ว่าวิธีการเรียนรู้เชิงลึกหลายอย่างเป็นสิ่งประดิษฐ์ใหม่ๆ แต่แนวคิดหลักเบื้องหลังการเรียนรู้จากข้อมูลได้รับการศึกษาเป็นเวลาหลายศตวรรษ ในความเป็นจริง มนุษย์มีความปรารถนาที่จะวิเคราะห์ข้อมูลและคาดการณ์ผลลัพธ์ในอนาคตมาช้านาน และความปรารถนานี้เป็นรากฐานของวิทยาศาสตร์ธรรมชาติและคณิตศาสตร์ส่วนใหญ่ สองตัวอย่างคือการแจกแจงแบบ Bernoulli ตั้งชื่อตาม Jacob Bernoulli (1655–1705) 23 และการแจกแจงแบบ Gaussian ที่ค้นพบโดย Carl Friedrich Gauss (1777–1855) 24 Gauss คิดค้นอัลกอริธึมแบบ least mean squares ซึ่งยังคงใช้ในปัจจุบันสำหรับปัญหาต่างๆ มากมายตั้งแต่การคำนวณประกันภัยไปจนถึงการวินิจฉัยทางการแพทย์ เครื่องมือดังกล่าวช่วยเสริมแนวทางการทดลองในวิทยาศาสตร์ธรรมชาติ—ตัวอย่างเช่น กฎหมายของโอห์มที่เกี่ยวข้องกับกระแสและแรงดันไฟฟ้าในตัวต้านทานอธิบายได้อย่างสมบูรณ์โดยโมเดลเชิงเส้น

แม้กระทั่งในยุคกลาง นักคณิตศาสตร์ก็มีความเข้าใจอย่างถ่องแท้เกี่ยวกับประมาณการ ตัวอย่างเช่น หนังสือเรขาคณิตของ Jacob Köbel (1460–1533) 25 แสดงให้เห็นถึงการเฉลี่ยความยาวของเท้าผู้ชายผู้ใหญ่ 16 คนเพื่อประเมินความยาวเท้าทั่วไปในประชากร (รูปที่ 1.4.1)

ขณะที่กลุ่มบุคคลออกจากโบสถ์ มีผู้ชายผู้ใหญ่ 16 คนถูกขอให้ยืนเรียงแถวและวัดเท้า จากนั้นนำผลรวมของการวัดเหล่านี้ไปหารด้วย 16 เพื่อรับประมาณการสำหรับสิ่งที่ตอนนี้เรียกว่าหนึ่งฟุต "อัลกอริธึม" นี้ได้รับการปรับปรุงในภายหลังเพื่อจัดการกับเท้าที่ผิดรูป; ชายสองคนที่มีเท้าสั้นที่สุดและยาวที่สุดถูกส่งไป ทำให้เฉลี่ยเฉพาะส่วนที่เหลือเท่านั้น นี่เป็นหนึ่งในตัวอย่างแรกๆ ของการประมาณค่าแบบตัดแต่ง

สถิติเริ่มต้นขึ้นจริง ๆ ด้วยความพร้อมใช้งานและการรวบรวมข้อมูล หนึ่งในผู้บุกเบิก Ronald Fisher (1890–1962) 26 ได้มีส่วนสำคัญในการพัฒนาแนวคิดและการประยุกต์ใช้ในด้านพันธุกรรม อัลกอริธึมจำนวนมากของเขา (เช่น การวิเคราะห์แบบแยกแยะเชิงเส้น) และแนวคิด (เช่น เมทริกซ์ข้อมูล Fisher) ยังคงมีบทบาทสำคัญในรากฐานของสถิติสมัยใหม่ แม้แต่ทรัพยากรข้อมูลของเขาก็ส่งผลกระทบอย่างยั่งยืน ชุดข้อมูล Iris ที่ Fisher เผยแพร่ในปี 1936 ยังคงใช้เพื่อสาธิตอัลกอริธึมการเรียนรู้ของเครื่องบางครั้ง Fisher ยังเป็นผู้สนับสนุนแนวคิดเรื่องการปรับปรุงพันธุ์มนุษย์ ซึ่งควรเตือนเราว่าศีลธรรมของการเรียนรู้ของเครื่องนั้นสำคัญเพียงใด

**Conclusion**

The field of machine learning is built on a foundation of centuries of research in mathematics, statistics, and computer science. As the field continues to develop, it is important to keep in mind the historical context and the ethical implications of our work.

โปรดทราบว่าหน่วยความจำแบบเข้าถึงโดยสุ่ม (RAM) ไม่ได้เพิ่มขึ้นในอัตราเดียวกับการเติบโตของข้อมูล ในขณะเดียวกัน การเพิ่มขึ้นของพลังการประมวลผลก็เกินหน้าการเติบโตของชุดข้อมูลเช่นกัน สิ่งนี้หมายความว่าโมเดลทางสถิติจำเป็นต้องมีประสิทธิภาพในการใช้หน่วยความจำมากขึ้น เพื่อให้สามารถใช้เวลาในการประมวลผลเพื่อปรับแต่งพารามิเตอร์ได้มากขึ้น เนื่องจากมีงบประมาณในการประมวลผลที่เพิ่มขึ้น

ดังนั้น จุดที่เหมาะสมที่สุดในการเรียนรู้ของเครื่องและสถิติจึงเปลี่ยนจากโมเดลเชิงเส้น (generalized) และวิธีเคอร์เนลไปเป็นเครือข่ายประสาทเทียมลึก นี่ก็เป็นหนึ่งในเหตุผลที่เสาหลักหลายประการของการเรียนรู้เชิงลึก เช่น เพอร์เซ็ปตรอนหลายชั้น (multilayer perceptrons) เครือข่ายประสาทเทียมแบบผันแปร (convolutional neural networks) หน่วยความจำระยะยาวระยะสั้น (long short-term memory) และการเรียนรู้แบบ Q (Q-Learning) ได้รับการ "ค้นพบใหม่" อีกครั้งในทศวรรษที่ผ่านมา หลังจากที่ซบเซาไปนานพอสมควร

ความก้าวหน้าล่าสุดในโมเดลทางสถิติ แอปพลิเคชัน และอัลกอริทึมบางครั้งถูกเปรียบเทียบกับการระเบิดแคมเบรียน ซึ่งเป็นช่วงเวลาแห่งความก้าวหน้าอย่างรวดเร็วในวิวัฒนาการของสปีชีส์ ในความเป็นจริง สถานะของศิลปะไม่ได้เป็นเพียงแค่ผลลัพธ์ของทรัพยากรที่มีอยู่ที่นำไปใช้กับอัลกอริทึมที่ใช้กันมาหลายสิบปี โปรดทราบว่ารายการแนวคิดด้านล่างนี้เป็นเพียงส่วนเล็กๆ ของสิ่งที่ช่วยให้ผู้วิจัยประสบความสำเร็จอย่างมากมายในช่วงทศวรรษที่ผ่านมา

* วิธีการใหม่ๆ สำหรับการควบคุมความจุ เช่น โดรปเอาต์ (dropout) ได้ช่วยลดปัญหาการโอเวอร์ฟิตติ้ง (overfitting) ซึ่งวิธีนี้จะทำการฉีดสัญญาณรบกวน (noise) เข้าไปในเครือข่ายประสาทเทียมระหว่างการฝึกอบรม
* กลไกการใส่ใจ (attention mechanisms) ได้แก้ปัญหาที่สองซึ่งเป็นปัญหาที่รบกวนสถิติมาเป็นเวลานานกว่าศตวรรษ นั่นคือ วิธีเพิ่มหน่วยความจำและความซับซ้อนของระบบโดยไม่เพิ่มจำนวนพารามิเตอร์ที่สามารถเรียนรู้ได้ ผู้วิจัยได้พบวิธีแก้ปัญหาที่ชาญฉลาดโดยใช้สิ่งที่มองได้ว่าเป็นโครงสร้างพอยนเตอร์ที่สามารถเรียนรู้ได้ (learnable pointer structure) แทนที่จะต้องจดจำลำดับข้อความทั้งหมด เช่น สำหรับการแปลภาษาเครื่องในรูปแบบที่มีมิติคงที่ สิ่งที่จำเป็นต้องจัดเก็บคือพอยนเตอร์ที่ชี้ไปยังสถานะกลางของกระบวนการแปลเท่านั้น สิ่งนี้ทำให้สามารถเพิ่มความแม่นยำได้อย่างมากสำหรับลำดับที่ยาว เนื่องจากโมเดลไม่จำเป็นต้องจดจำลำดับทั้งหมดก่อนที่จะเริ่มสร้างลำดับใหม่
* สถาปัตยกรรม Transformer (Vaswani et al., 2017) ซึ่งสร้างขึ้นบนพื้นฐานของกลไกการใส่ใจเพียงอย่างเดียว ได้รับการแนะนำในปี 2017 และกลายเป็นสถานะของศิลปะอย่างรวดเร็วสำหรับงานประมวลผลภาษาธรรมชาติหลายประเภท

โดยสรุป ความก้าวหน้าล่าสุดในการเรียนรู้เชิงลึกเกิดจากปัจจัยหลายประการ รวมถึงการมีข้อมูลจำนวนมหาศาล การเพิ่มขึ้นของพลังการประมวลผล และความก้าวหน้าของอัลกอริทึมใหม่



การเรียนรู้เชิงลึกได้พัฒนาไปอย่างมากในช่วงทศวรรษที่ผ่านมา โดยได้รับแรงหนุนจากปัจจัยหลายประการ รวมถึงการที่มีข้อมูลจำนวนมหาศาล พลังการประมวลผลที่เพิ่มขึ้น และความก้าวหน้าของอัลกอริทึม

หนึ่งในการพัฒนาที่สำคัญที่สุดคือการถือกำเนิดของสถาปัตยกรรม Transformer ซึ่งได้กลายเป็นมาตรฐานสำหรับการประมวลผลภาษาธรรมชาติ การมองเห็นด้วยคอมพิวเตอร์ การรู้จำเสียง และการเรียนรู้แบบเสริมแรง Transformer ได้พิสูจน์ให้เห็นแล้วว่ามีประสิทธิภาพเหนือกว่าสถาปัตยกรรมอื่นๆ ในการจัดการกับชุดข้อมูลขนาดใหญ่และโมเดลขนาดใหญ่ นอกจากนี้ยังมีความสำเร็จในการทำงานที่หลากหลาย เช่น การเล่นเกม Atari การอธิบายภาพ การแชท และการควบคุมหุ่นยนต์

อีกก้าวสำคัญคือการพัฒนาของโมเดลภาษาขนาดใหญ่ (LLMs) LLMs ได้รับการฝึกฝนบนชุดข้อมูลขนาดมหาศาลของข้อความและรหัส ซึ่งทำให้พวกมันสามารถทำหน้าที่ที่หลากหลาย เช่น การสร้างข้อความ การแปลภาษา และการเขียนโค้ด LLMs ยังสามารถใช้ในการตอบคำถามที่ซับซ้อนและการแก้ปัญหาในโลกแห่งความจริงได้อีกด้วย

ความก้าวหน้าในด้านการเรียนรู้เชิงลึกยังนำไปสู่การพัฒนาเทคนิคการสร้างแบบจำลองใหม่ๆ เช่น generative adversarial networks (GANs) และ diffusion models GANs ใช้สองโมเดลประสาทเทียมเพื่อแข่งขันกันในการสร้างข้อมูลที่สมจริง ในขณะที่ diffusion models ใช้เทคนิคการแพร่กระจายเพื่อสร้างข้อมูลจากสัญญาณรบกวน

ความก้าวหน้าเหล่านี้ได้เปิดใช้งานแอปพลิเคชันใหม่ๆ มากมายในการเรียนรู้เชิงลึก เช่น การสร้างศิลปะดิจิทัล การออกแบบผลิตภัณฑ์ใหม่ และการพัฒนายารักษาโรคใหม่ๆ

โดยรวมแล้ว การเรียนรู้เชิงลึกได้กลายเป็นเทคโนโลยีที่ทรงพลังที่สุดอย่างหนึ่งในปัจจุบัน และกำลังปฏิวัติอุตสาหกรรมและสาขาต่าง ๆ มากมาย

#1.6 Success Stories




ปัญญาประดิษฐ์ (AI) มีประวัติอันยาวนานในการให้ผลลัพธ์ที่ยากจะทำได้ด้วยวิธีอื่น เช่น ระบบการเรียงลำดับจดหมายโดยใช้การจดจำอักขระด้วยแสง (OCR) ได้รับการนำไปใช้มาตั้งแต่ทศวรรษ 1990 นี่คือที่มาของชุดข้อมูล MNIST ที่มีชื่อเสียงของตัวเลขที่เขียนด้วยลายมือ สิ่งเดียวกันนี้ใช้กับการอ่านเช็คเพื่อฝากเงินเข้าบัญชีและการให้คะแนนความน่าเชื่อถือทางเครดิตของผู้สมัคร ธุรกรรมทางการเงินได้รับการตรวจสอบหาการฉ้อโกงโดยอัตโนมัติ





ปัญญาประดิษฐ์ (AI) ได้รับความสนใจอย่างมากในช่วงไม่กี่ปีที่ผ่านมา ส่วนใหญ่เป็นเพราะโซลูชันสำหรับปัญหาที่เคยถือว่าเป็นปัญหาที่ยากต่อการแก้ไขและเกี่ยวข้องโดยตรงกับผู้บริโภค ความก้าวหน้าหลายประการเหล่านี้เกิดจากการเรียนรู้เชิงลึก (deep learning)

* **ผู้ช่วยอัจฉริยะ** เช่น Siri ของ Apple, Alexa ของ Amazon และผู้ช่วยของ Google สามารถตอบสนองต่อคำขอที่พูดได้ด้วยความแม่นยำในระดับที่เหมาะสม ซึ่งรวมถึงงานที่ไม่ซับซ้อน เช่น การเปิดสวิตช์ไฟ และงานที่ซับซ้อนกว่า เช่น การจองคิวตัดผมและให้บริการสนับสนุนทางโทรศัพท์ นี่อาจเป็นสัญญาณที่สังเกตได้เห็นมากที่สุดว่า AI กำลังส่งผลกระทบต่อชีวิตของเรา
* ส่วนประกอบสำคัญของผู้ช่วยอัจฉริยะคือความสามารถในการจดจำเสียงได้อย่างแม่นยำ ความแม่นยำของระบบดังกล่าวได้เพิ่มขึ้นอย่างค่อยเป็นค่อยไปจนถึงจุดที่บรรลุความเท่าเทียมกับมนุษย์สำหรับบางแอปพลิเคชัน (Xiong et al., 2018)
* **การจดจำวัตถุ** ก็เช่นกันที่ก้าวไกลมามาก การระบุวัตถุในภาพเป็นงานที่ค่อนข้างท้าทายในปี 2010 ในเกณฑ์มาตรฐาน ImageNet นักวิจัยจาก NEC Labs และ University of Illinois at Urbana-Champaign ได้รับอัตราข้อผิดพลาดสูงสุดห้าข้อที่ 28% (Lin et al., 2010) ภายในปี 2017 อัตราข้อผิดพลาดนี้ลดลงเหลือ 2.25% (Hu et al., 2018) ผลลัพธ์ที่น่าทึ่งที่คล้ายกันนี้ได้ถูกนำไปใช้กับการระบุเสียงนกร้องและการวินิจฉัยโรคมะเร็งผิวหนัง
* ความสามารถในการเล่นเกมเคยใช้เป็นตัววัดความสามารถของมนุษย์ ตั้งแต่ TD-Gammon โปรแกรมสำหรับเล่นเกมแบ็กแกมมอนโดยใช้การเรียนรู้เสริมด้วยความแตกต่างของเวลารายการ การพัฒนาอัลกอริธึมและการคำนวณได้นำไปสู่การพัฒนาอัลกอริธึมสำหรับแอปพลิเคชันที่หลากหลาย เมื่อเทียบกับแบ็กแกมมอนแล้ว หมากรุกมีสถานะที่ซับซ้อนกว่ามากและชุดของการกระทำ DeepBlue เอาชนะ Garry Kasparov โดยใช้ความขนานขนาดใหญ่ ฮาร์ดแวร์เฉพาะทาง และการค้นหาที่มีประสิทธิภาพผ่านต้นไม้เกม (Campbell et al., 2002) Go นั้นยากกว่าเดิม เนื่องจากมีสถานะที่มากมาย AlphaGo บรรลุความเท่าเทียมกับมนุษย์ในปี 2015 โดยใช้การเรียนรู้เชิงลึกควบคู่กับการสุ่มตัวอย่างต้นไม้ Monte Carlo (Silver et al., 2016) ความท้าทายใน Poker คือสถานะนั้นมีขนาดใหญ่และสังเกตได้เพียงบางส่วน (เราไม่รู้ไพ่ของฝ่ายตรงข้าม) Libratus เกินประสิทธิภาพของมนุษย์ใน Poker โดยใช้กลยุทธ์ที่มีโครงสร้างอย่างมีประสิทธิภาพ (Brown and Sandholm, 2017)
* อีกหนึ่งตัวบ่งชี้ความก้าวหน้าของ AI คือการถือกำเนิดของยานพาหนะขับเคลื่อนอัตโนมัติ ในขณะที่การขับขี่อัตโนมัติอย่างเต็มรูปแบบยังไม่สามารถเข้าถึงได้ แต่ก็มีความก้าวหน้าที่ดีเยี่ยมในทิศทางนี้ โดยบริษัทต่างๆ เช่น Tesla, NVIDIA และ Waymo จัดส่งผลิตภัณฑ์ที่ช่วยให้ขับขี่อัตโนมัติได้บางส่วน สิ่งที่ทำให้การขับขี่อัตโนมัติอย่างเต็มรูปแบบท้าทายคือการขับขี่ที่เหมาะสมจำเป็นต้องมีความสามารถในการรับรู้ เหตุผล และรวมกฎเกณฑ์เข้ากับระบบ ขณะนี้ การเรียนรู้เชิงลึกถูกใช้เป็นหลักในด้านภาพของปัญหาเหล่านี้ ส่วนที่เหลือได้รับการปรับแต่งอย่างหนักโดยวิศวกร





นี่เป็นเพียงแค่ส่วนหนึ่งของการประยุกต์ใช้การเรียนรู้ของเครื่องที่สำคัญเท่านั้น เช่น หุ่นยนต์ โลจิสติก ชีววิทยาคำนวณ ฟิสิกส์อนุภาค และดาราศาสตร์ต่างก็เป็นหนี้ความก้าวหน้าที่น่าประทับใจที่สุดในปัจจุบันบางส่วนให้กับการเรียนรู้ของเครื่อง ซึ่งกำลังกลายเป็นเครื่องมือที่แพร่หลายสำหรับวิศวกรและนักวิทยาศาสตร์

บ่อยครั้ง คำถามเกี่ยวกับวันสิ้นโลกของ AI และความเป็นไปได้ของจุดเอกพจน์ (singularity) ถูกหยิบยกขึ้นมาในบทความที่ไม่ใช่ด้านเทคนิค ความกลัวคือระบบการเรียนรู้ของเครื่องจะกลายเป็นสิ่งมีชีวิตและตัดสินใจโดยอิสระจากโปรแกรมเมอร์ซึ่งส่งผลโดยตรงต่อชีวิตของมนุษย์ ในระดับหนึ่ง AI ได้มีผลกระทบต่อการดำรงชีวิตของมนุษย์โดยตรงอยู่แล้ว: ความน่าเชื่อถือทางเครดิตได้รับการประเมินโดยอัตโนมัติ เครื่องบินนำร่องส่วนใหญ่เป็นผู้นำทางยานพาหนะ การตัดสินใจเกี่ยวกับการให้ประกันตัวใช้ข้อมูลทางสถิติเป็นข้อมูลเบื้องต้น ยิ่งไปกว่านั้น เราสามารถขอให้ Alexa เปิดเครื่องชงกาแฟได้

โชคดีที่เรายังห่างไกลจากระบบ AI ที่มีสติสัมปชัญญะซึ่งสามารถควบคุมผู้สร้างมนุษย์อย่างจงใจได้ ประการแรก ระบบ AI ถูกออกแบบ ฝึกอบรม และใช้งานในลักษณะที่เฉพาะเจาะจงและมุ่งเป้าไปที่เป้าหมาย แม้ว่าพฤติกรรมของพวกเขาอาจให้ภาพลวงตาของสติปัญญาโดยทั่วไป แต่ก็เป็นการผสมผสานกฎ กฎเบื้องต้น และโมเดลทางสถิติที่อยู่เบื้องหลังการออกแบบ ประการที่สอง ขณะนี้ไม่มีเครื่องมือสำหรับปัญญาประดิษฐ์ทั่วไปที่สามารถปรับปรุงตัวเอง เหตุผลเกี่ยวกับตัวเอง และสามารถปรับเปลี่ยน ขยาย และปรับปรุงสถาปัตยกรรมของตัวเองในขณะที่พยายามแก้ปัญหาทั่วไป

ความกังวลเร่งด่วนกว่ามากคือวิธีที่ AI ถูกนำมาใช้ในชีวิตประจำวันของเรา เป็นไปได้ว่างานประจำหลายอย่างที่มนุษย์ทำอยู่ปัจจุบันสามารถและจะถูกทำให้เป็นอัตโนมัติ หุ่นยนต์ในฟาร์มน่าจะช่วยลดต้นทุนสำหรับเกษตรกรอินทรีย์ แต่พวกมันจะทำให้การเก็บเกี่ยวเป็นอัตโนมัติด้วยเช่นกัน เฟสนี้ของการปฏิวัติอุตสาหกรรมอาจส่งผลกระทบอย่างลึกซึ้งต่อสังคมจำนวนมาก เนื่องจากงานประจำให้การจ้างงานเป็นจำนวนมากในหลายประเทศ นอกจากนี้ โมเดลทางสถิติเมื่อนำไปใช้โดยไม่ระมัดระวังอาจนำไปสู่ความลำเอียงด้านเชื้อชาติ เพศ หรืออายุ และเพิ่มความกังวลที่สมเหตุสมผลเกี่ยวกับความยุติธรรมตามขั้นตอนหากนำไปใช้เพื่อขับเคลื่อนการตัดสินใจที่มีผลลัพธ์สำคัญ เป็นสิ่งสำคัญที่จะต้องแน่ใจว่าอัลกอริทึมเหล่านี้ถูกใช้ด้วยความระมัดระวัง จากสิ่งที่เรารู้ในปัจจุบัน นี่เป็นความกังวลเร่งด่วนกว่ามากสำหรับเรา มากกว่าศักยภาพของปัญญาประดิษฐ์ที่เหนือชั้นที่จะทำลายมนุษยชาติ

**สรุป**

การเรียนรู้ของเครื่องกำลังกลายเป็นเครื่องมือที่ทรงพลังสำหรับการแก้ไขปัญหาที่ซับซ้อนหลากหลาย อย่างไรก็ตาม สิ่งสำคัญคือต้องตระหนักถึงความท้าทายที่อาจเกิดขึ้น เช่น ความลำเอียงทางสังคมและผลกระทบต่อการจ้างงาน เราจำเป็นต้องใช้เทคโนโลยีนี้อย่างรับผิดชอบและเป็นธรรม เพื่อให้มั่นใจว่าทุกคนได้รับประโยชน์จากมัน

#1.7 The Essence of Deep Learning



จนถึงตอนนี้ เราได้พูดคุยกันเกี่ยวกับการเรียนรู้ของเครื่องในแง่กว้าง การเรียนรู้เชิงลึก (deep learning) เป็นส่วนหนึ่งของการเรียนรู้ของเครื่องที่เกี่ยวข้องกับโมเดลที่ใช้เครือข่ายประสาทเทียมหลายชั้น มันลึกในแง่ที่โมเดลของมันเรียนรู้การแปลงหลายชั้น ในขณะที่สิ่งนี้อาจฟังดูแคบ แต่การเรียนรู้เชิงลึกได้ก่อให้เกิดโมเดล เทคนิค การกำหนดสูตรปัญหา และแอปพลิเคชันมากมาย ความเข้าใจมากมายได้รับการพัฒนาเพื่ออธิบายประโยชน์ของความลึก

อาจกล่าวได้ว่าการเรียนรู้ของเครื่องทั้งหมดมีการคำนวณหลายชั้น โดยชั้นแรกประกอบด้วยขั้นตอนการประมวลผลฟีเจอร์ สิ่งที่ทำให้การเรียนรู้เชิงลึกแตกต่างคือการดำเนินการที่เรียนรู้ในแต่ละชั้นของตัวแทนนั้นจะถูกเรียนรู้ร่วมกันจากข้อมูล ปัญหาที่เราได้กล่าวถึงไปแล้ว เช่น การเรียนรู้จากสัญญาณเสียงดิบ ค่าพิกเซลดิบของภาพ หรือการแม็ประหว่างประโยคที่มีความยาวโดยพลการกับประโยคในภาษาต่างประเทศ เป็นปัญหที่การเรียนรู้เชิงลึกมีความโดดเด่นและวิธีการแบบดั้งเดิมล้มเหลว ปรากฎว่าโมเดลหลายชั้นเหล่านี้สามารถจัดการกับข้อมูลการรับรู้ระดับต่ำในลักษณะที่เครื่องมือก่อนหน้านี้ไม่สามารถทำได้

ในบรรดาแนวทางการเรียนรู้เชิงลึก จุดร่วมที่สำคัญที่สุดน่าจะเป็นการฝึกอบรมแบบครบวงจร นั่นคือ แทนที่จะประกอบระบบจากคอมโพเนนต์ที่ปรับแต่งทีละรายการ เราจะสร้างระบบและปรับแต่งประสิทธิภาพร่วมกัน ตัวอย่างเช่น ในด้านการคอมพิวเตอร์วิทัศน์ นักวิทยาศาสตร์เคยแยกกระบวนการวิศวกรรมฟีเจอร์ออกจากกระบวนการสร้างโมเดลการเรียนรู้ของเครื่อง ตัวตรวจจับขอบแบบ Canny (Canny, 1987) และตัวดึงฟีเจอร์ SIFT ของ Lowe (Lowe, 2004) เป็นราชาแห่งวงการมานานกว่าทศวรรษในฐานะอัลกอริทึมสำหรับการแม็ปภาพเป็นเวกเตอร์ฟีเจอร์ ในอดีต ส่วนสำคัญของการใช้การเรียนรู้ของเครื่องกับปัญหาเหล่านี้ประกอบด้วยการหาวิธีการแปลงข้อมูลด้วยตนเองให้เป็นรูปแบบที่เหมาะสมกับโมเดลตื้น น่าเสียดายที่สิ่งที่มนุษย์สามารถทำได้ด้วยความคิดสร้างสรรค์นั้นมีจำกัดเมื่อเทียบกับการประเมินอย่างสม่ำเสมอเกี่ยวกับตัวเลือกหลายล้านรายการที่ดำเนินการโดยอัลกอริทึมโดยอัตโนมัติ เมื่อการเรียนรู้เชิงลึกเข้ามาแทนที่ ตัวดึงฟีเจอร์เหล่านี้ถูกแทนที่ด้วยฟิลเตอร์ที่ปรับแต่งโดยอัตโนมัติ ซึ่งให้ความแม่นยำที่สูงขึ้น

ดังนั้น ข้อดีสำคัญประการหนึ่งของการเรียนรู้เชิงลึกคือมันไม่เพียงแต่แทนที่โมเดลตื้นที่ส่วนท้ายของกระบวนการเรียนรู้แบบเดิมเท่านั้น แต่ยังแทนที่กระบวนการวิศวกรรมฟีเจอร์ที่ใช้แรงงานมากด้วย นอกจากนี้ ด้วยการแทนที่การประมวลผลก่อนหน้าเฉพาะโดเมนจำนวนมาก การเรียนรู้เชิงลึกได้กำจัดขอบเขตหลายประการที่เคยแยกการมองเห็นด้วยคอมพิวเตอร์ การจดจำเสียง การประมวลผลภาษาธรรมชาติ การแพทย์สารสนเทศ และสาขาการใช้งานอื่นๆ ออกจากกัน ดังนั้น การเรียนรู้เชิงลึกจึงเสนอชุดเครื่องมือที่เป็นหนึ่งเดียวสำหรับการแก้ปัญหาที่หลากหลาย

นอกเหนือจากการฝึกอบรมแบบครบวงจรแล้ว เรายังกำลังประสบกับการเปลี่ยนแปลงจากคำอธิบายทางสถิติเชิงพารามิเตอร์เป็นโมเดลที่ไม่ใช่พารามิเตอร์อย่างสมบูรณ์ เมื่อข้อมูลขาดแคลน เราจำเป็นต้องพึ่งพาสมมติฐานที่ง่ายขึ้นเกี่ยวกับความเป็นจริงเพื่อให้ได้โมเดลที่มีประโยชน์ เมื่อข้อมูลมีมากมาย สิ่งเหล่านี้สามารถแทนที่ด้วยโมเดลที่ไม่ใช่พารามิเตอร์ที่เหมาะสมกับข้อมูล

#1.8 Summary



การเรียนรู้ของเครื่อง (Machine learning) เป็นการศึกษาว่าระบบคอมพิวเตอร์สามารถใช้ประโยชน์จากประสบการณ์ (บ่อยครั้งคือข้อมูล) เพื่อปรับปรุงประสิทธิภาพในการทำงานเฉพาะได้อย่างไร มันผสมผสานแนวคิดจากสถิติ การขุดค้นข้อมูล และการเพิ่มประสิทธิภาพ โดยส่วนใหญ่จะใช้เป็นวิธีการนำโซลูชัน AI ไปใช้

การเรียนรู้แบบแสดงผล (Representational learning) เป็นประเภทหนึ่งของการเรียนรู้ของเครื่องที่มุ่งเน้นไปที่วิธีค้นหาแนวทางที่เหมาะสมในการแสดงข้อมูลโดยอัตโนมัติ การเรียนรู้เชิงลึก (Deep learning) ซึ่งถือเป็นการเรียนรู้แบบแสดงผลหลายชั้นผ่านการเรียนรู้การแปลงหลายชั้นนั้น ไม่เพียงแต่แทนที่โมเดลตื้นที่ส่วนท้ายของกระบวนการเรียนรู้ของเครื่องแบบเดิมเท่านั้น แต่ยังแทนที่กระบวนการวิศวกรรมฟีเจอร์ที่ใช้แรงงานมากด้วย ความก้าวหน้าล่าสุดส่วนใหญ่ในการเรียนรู้เชิงลึกนั้นเกิดจากข้อมูลจำนวนมากที่เกิดจากเซ็นเซอร์ราคาถูกและแอปพลิเคชันในระดับอินเทอร์เน็ต รวมถึงความก้าวหน้าอย่างมากในการคำนวณ โดยส่วนใหญ่เกิดจาก GPU นอกจากนี้ การมีเฟรมเวิร์กการเรียนรู้เชิงลึกที่ทรงประสิทธิภาพยังทำให้การออกแบบและการนำไปใช้การเพิ่มประสิทธิภาพระบบทั้งหมดง่ายขึ้นอย่างมาก ซึ่งเป็นส่วนประกอบสำคัญในการรับประสิทธิภาพสูง

**สรุป**

การเรียนรู้เชิงลึกเป็นเทคนิคการเรียนรู้ของเครื่องที่ทรงพลังซึ่งสามารถใช้ในการแก้ปัญหาที่หลากหลาย ความก้าวหน้าล่าสุดส่วนใหญ่ในการเรียนรู้เชิงลึกนั้นเกิดจากข้อมูลจำนวนมากและความก้าวหน้าในการคำนวณ การเรียนรู้เชิงลึกมีศักยภาพที่จะปฏิวัติหลายอุตสาหกรรมและสาขาวิชา

วิธีหนึ่งในการปรับปรุงรหัสของฉันคือการใช้การเรียนรู้ของเครื่องเพื่อเรียนรู้พฤติกรรมที่ต้องการ ตัวอย่างเช่น ฉันสามารถฝึกโมเดลเพื่อทำนายไฮเปอร์พารามิเตอร์ที่ดีที่สุดสำหรับโมเดลและชุดข้อมูลที่กำหนด ซึ่งจะช่วยให้ฉันกำหนดตัวเลือกการออกแบบที่ดีที่สุดโดยอัตโนมัติ แม้สำหรับปัญหาที่ฉันไม่เคยเห็นมาก่อน

ในการเรียนรู้พฤติกรรมที่ต้องการ ฉันจะต้องมีชุดข้อมูลตัวอย่างของโมเดลการเรียนรู้ของเครื่อง ไฮเปอร์พารามิเตอร์ และประสิทธิภาพของโมเดลเหล่านั้นบนชุดข้อมูลต่างๆ ชุดข้อมูลนี้จะต้องมีขนาดใหญ่และเป็นตัวแทนของประเภทของปัญหาที่ฉันต้องการแก้ไข



2. ปัญหาใดที่ฉันพบว่ามีตัวอย่างมากมายสำหรับการแก้ปัญหา แต่ยังไม่มีวิธีเฉพาะเจาะจงในการทำให้เป็นระบบอัตโนมัติ ปัญหาเหล่านี้อาจเป็นผู้สมัครที่เหมาะสมสำหรับการใช้การเรียนรู้เชิงลึก

ตอบ ตัวอย่างหนึ่งของปัญหาที่มีตัวอย่างมากมายสำหรับการแก้ปัญหา แต่ยังไม่มีวิธีเฉพาะเจาะจงในการทำให้เป็นระบบอัตโนมัติคือปัญหาการแปลข้อความระหว่างภาษา มีระบบการแปลภาษาเครื่องมากมายที่มีอยู่แล้ว แต่ทั้งหมดพึ่งพากฎและฮิวริสติกที่สร้างด้วยมือ ระบบเหล่านี้มักไม่แม่นยำและยากต่อการบำรุงรักษา

การเรียนรู้เชิงลึกเป็นแนวทางที่มีแนวโน้มดีสำหรับการแปลภาษาเครื่องเนื่องจากสามารถเรียนรู้ที่จะแปลข้อความได้โดยตรงจากตัวอย่าง ซึ่งหมายความว่าระบบการเรียนรู้เชิงลึกสามารถฝึกอบรมได้กับชุดข้อมูลขนาดใหญ่ของข้อความที่แปลแล้ว และจากนั้นสามารถใช้เพื่อแปลข้อความใหม่ได้โดยไม่จำเป็นต้องใช้กฎหรือฮิวริสติกที่สร้างด้วยมือใดๆ

อีกตัวอย่างของปัญหาที่เหมาะสมกับการเรียนรู้เชิงลึกคือปัญหาการจดจำภาพ มีระบบการจดจำภาพที่มีอยู่มากมาย แต่ทั้งหมดพึ่งพาคุณสมบัติที่สร้างด้วยมือ ระบบเหล่านี้มักไม่แม่นยำและฝึกอบรมได้ยาก

การเรียนรู้เชิงลึกสามารถใช้เพื่อเรียนรู้คุณสมบัติของภาพได้โดยตรงจากข้อมูล ซึ่งหมายความว่าระบบการเรียนรู้เชิงลึกสามารถฝึกอบรมได้กับชุดข้อมูลขนาดใหญ่ของภาพที่ติดฉลาก และจากนั้นสามารถใช้เพื่อจดจำวัตถุในภาพใหม่ได้โดยไม่จำเป็นต้องใช้คุณสมบัติที่สร้างด้วยมือใดๆ

อัลกอริทึม ข้อมูล และการคำนวณนั้นเชื่อมโยงกันหมด การเลือกอัลกอริทึมขึ้นอยู่กับลักษณะเฉพาะของข้อมูลและทรัพยากรการคำนวณที่มีอยู่

ตัวอย่างเช่น อัลกอริทึมบางอย่างมีประสิทธิภาพมากกว่าอัลกอริทึมอื่นๆ แต่ต้องใช้ข้อมูลมากขึ้นในการฝึก อัลกอริทึมอื่นๆ มีประสิทธิภาพน้อยกว่า แต่สามารถเรียนรู้ได้จากข้อมูลน้อยลง

ทรัพยากรการคำนวณที่มีอยู่ยังมีบทบาทในการเลือกอัลกอริทึม อัลกอริทึมบางอย่างต้องใช้พลังการคำนวณมาก ในขณะที่อัลกอริทึมอื่นๆ สามารถทำงานบนแล็ปท็อป ดยทั่วไปแล้ว อัลกอริทึมที่ดีที่สุดสำหรับปัญหาที่กำหนดคืออัลกอริทึมที่สามารถใช้ฝึกบนข้อมูลที่มีอยู่และสามารถทำงานบนทรัพยากรการคำนวณที่มีอยู่

4. ตั้งชื่อการตั้งค่าบางประการที่การฝึกอบรมแบบ end-to-end ไม่ใช่แนวทางเริ่มต้น แต่อาจมีประโยชน์(Name some settings where end-to-end training is not currently the default approach but
where it might be useful.
)

การตั้งค่าบางประการที่การฝึกอบรมแบบ end-to-end ไม่ใช่แนวทางเริ่มต้น แต่อาจมีประโยชน์ ได้แก่:

การประมวลผลภาษาธรรมชาติ (NLP) โดยทั่วไปแล้ว NLP จะดำเนินการโดยใช้ชุดของอัลกอริทึมที่แตกต่างกันเพื่อทำงานต่างๆ เช่น การแปล การทำข้อมูลสรุป และตอบคำถาม การฝึกอบรมแบบ end-to-end สามารถใช้ในการฝึกโมเดลเดียวเพื่อทำงานทั้งหมดในชุดข้อมูล NLP ซึ่งจะช่วยลดจำนวนอัลกอริทึมและข้อมูลที่จำเป็นในการดำเนินการงานเหล่านี้
หุ่นยนต์ หุ่นยนต์มักใช้เซ็นเซอร์ต่างๆ เพื่อรวบรวมข้อมูลเกี่ยวกับสภาพแวดล้อมของพวกมัน จากนั้นจะใช้ข้อมูลนี้เพื่อควบคุมการเคลื่อนไหว การฝึกอบรมแบบ end-to-end สามารถใช้ในการฝึกโมเดลเดียวเพื่อควบคุมการเคลื่อนไหวของหุ่นยนต์โดยตรงจากข้อมูลเซ็นเซอร์ ซึ่งจะช่วยลดจำนวนเซ็นเซอร์และข้อมูลที่จำเป็นในการควบคุมหุ่นยนต์
โดยทั่วไปแล้ว การฝึกอบรมแบบ end-to-end เป็นแนวทางที่มีแนวโน้มดีสำหรับปัญหาที่ซับซ้อนซึ่งต้องใช้ชุดของอัลกอริทึมหรือเซ็นเซอร์ในการทำงาน อย่างไรก็ตาม การฝึกอบรมแบบ end-to-end อาจเป็นเรื่องยากที่จะนำไปใช้และต้องใช้ข้อมูลจำนวนมาก
